## Motivation/Versuchsziel

Die Umwandlung von Wärme (z. B. hervorgehend aus Verbrennungen) in Mechanische Arbeit ist menscheitsgeschichtlich ein außerordentlich wichtiges Unterfangen. Die Dampfmaschienen des frühen 19. Jahrhunderts waren jedoch nicht sonderlich effizient und die hohen Drücke machten sie anfällig für Explosionen. Der 1816 patentierte Stirlingmotor löste diese Probleme mit einem (verhältnismäßig) besserem Wirkungsgrad und ohne Verwendung hoher Drücke. Dieser Heißluftmotor lässt sich sowohl als Wärmemaschiene, Kältemaschiene als auch als Wärmepumpe betreiben. Hier sollen die Effizienzwerte ermittelt und, wo möglich, die Verlustquellen benannt und quantisiert werden. 



## Grundlagen

### Der Heißluftmotor 

In diesem Verusch wird ein $\beta$-Heißluftmotor genutzt. Dieser besteht aus einem Arbeits- und einem  Verdrängerkolben. Der obere Bereich des Zylinders ist beheitzt  Es werden (im idealisierten Bild) zyklisch vier Zustände durchlaufen, wobei zwischen ihnen je isotherme oder isochore Prozesse liegen. Wir gehen diese entsprechend Abbildung 1 durch:


<center><b>Abbildung 1:  Aufbau des Stirlingmotors und pV-Diagramm (Quelle: Script)</b> </center>



1. **isotherme Expansion:** Der Verdrängerkolben befindet sich hier ganz unten um Zylinder, dem Totpunkt. Wärmt sich nun die Luft im oberen Teil, so dehnt sich diese aus und der Arbeitskolben wird hinuntergeschoben. Dabei wird die Wärme $Q_1$ aufgenommen und die Arbeit $W_1$ verrichtet.

2. **isochore Abkühlung:** Da nun Luft in den unteren Teil strömen kann, kühlt sich die Luft auf Temperatur $T_2$ unter abgabe von $Q_2$.

3. **isotherme Kompression:** Der Arbeitskolben verdichtet nun die Luft unter Aufwendung der Arbeit $W_3$ und Verlust der Wärme $Q_3$.

4. **isochore Erwärmung:** Der Verdrängungskolben bewegt sich nun nach unten, sodass die Luft sich im oberen Zylinderteil wieder auf $T_1$ erwärmt. 


Ein solcher idealer Stirlinprozess ist jedoch technisch nicht umsesetzbar, da die Kolben sich nicht diskontinuierlich bewegen lassen. Praktisch bewegen sich beide Kolben etwa sinusförmig, mit einer Phasenverschiebung von $\frac{\pi}{2}$. Damit überlagern sich die Vorgänge praktisch immer etwas gegenseitig.

### Der Regenerator

Um die Effizienz des Heißluftmotors weiter zu steigern, wird ein sog. Regenerator eingesetzt. Dies bedeutet, dass ein Teil der vom Gas abgegebenen Wärme zwischengespeichert und dem Gas hinterher wieder zugeführt werden kann. Im Versuch wird dies über Kupferwolle im Verdrängerkolben relaisiert, welche Wärme aufnimmt, wenn die warme Luft durch sie hindurch in den kälteren Bereich strömt und vice versa. 

### Thermodynamische Beschreibung

Es gelte zunächst der erste Hauptsatz der Thermodynamik
\begin{align}
dQ = dU + pdV
\end{align}
Eine Erhöhung der inneren Energie $U$ führt bei einem idealen Gas also zu einem Temperaturanstieg. Wir nutzen nun die molare Wärmekapazität  $C_V$ ein und erhalten
\begin{align}
dQ = C_V νdT + pdV.
\end{align}
wobei $\nu$ die Stoffmenge in Mol sei. Wir wollen nun den idealen Kreisprozess nachverfolgen und den Wirkungsgrad
\begin{align}
\eta=\frac{W_n}{Q^+}
\end{align}
bestimmen, wobei $W_n$ die nutzbare verrichtete Arbeit ist und $Q^+$ die investierte Wärme:
1. Hier verschwindet $dT$ offenbar, also lässt sich mit der idealen Gasgleichung und Integration schnell 
\begin{align}
Q_1=W_1=\nu RT_1 \ln (\frac{V_2}{V_1})
\end{align}
erhalten. 
2. Nun wird $pdV$ verschwinden. Die Temperaturänderugen ändern also direkt und außschließlich die Innere Energie. Es wird somit keine Arbeit verrichtet.
\begin{align}
W_2=0
\end{align}
analog zu oben findet sich
\begin{align}
Q_2=-C_V \nu (T_1-T_2) 
\end{align}
3. Analog zum ersten Schritt erhält man
\begin{align}
Q_3=W_3=-\nu RT_2 \ln (\frac{V_2}{V_1})
\end{align}

4. Analog zum zweiten Schritt folgt
\begin{align}
W_4=0
\end{align}
analog zu oben findet sich
\begin{align}
Q_4=C_V \nu (T_1-T_2) 
\end{align}


Zusammengefasst ergibt sich:
\begin{equation}
W_{N}=\oint p d V=W_{1}+W_{3}=\nu R\left(T_{1}-T_{2}\right) \ln \left(\frac{V_{2}}{V_{1}}\right)
\end{equation}
und damit schließlich
\begin{align}
	\eta = \frac{nR (T_1 - T_2) \ln\lr{\frac{V_{2}}{V_{1}}}}{ nRT_{1}\ln\lr{\frac{V_{2}}{V_{1}}} + nc_V(T_{1}-T_{2})}  = \frac{(T_1 - T_2) \ln\lr{\frac{V_{2}}{V_{1}}}}{T_{1}\ln\lr{\frac{V_{2}}{V_{1}}} + \frac{c_V}{R}(T_{1}-T_{2})}.
\end{align}
Wird ein (idealer) Regenerator eingesetzt, also $Q_4$ komplett durch $Q_2$ ausgeglichen, findet man:
\begin{align}
	\eta^R= \frac{(T_1 - T_2) \ln\lr{\frac{V_{2}}{V_{1}}}}{T_{1}\ln\lr{\frac{V_{2}}{V_{1}}}} = \frac{T_1-T_2}{T_1}
\end{align}
Dieser Wirkungsgrad ist der maximal Mögliche, der sogenannte Carnot-Wirkungsgrad. Wird der Sterlingmotor als Kältemaschiene oder Wärmepumpe betrieben, so invertiert sich der Wirkungsgrad:
\begin{align}
	\eta =\frac{Q_2}{W_n}=\frac{T_1}{T_1-T_2}
\end{align}
Wobei $Q_2$ die zugeführte Wärme ist.  Ist $Q_1$ die abgeführte Wärme, so ergibt sich für eine Wärmepumpe die Effizienz 
\begin{align}
	\epsilon = \frac{Q_1}{W_n} = \frac{T_1}{T_1-T_2} > 1 
\end{align}

## Auswertung

### Betrieb als Kältemaschine 

Wir wollen den Wirkungsgrad bestimmen. Dazu bestimmen wir zunächst die Leistung aus der Kompenationsmessung:
\begin{align}
	P_H&=U_H\cdot I_H\label{eq:P_H}\\
	\Delta P_H &= P_H \cdot \sqrt{\lr{\frac{\Delta U_H}{U_H}}^2+\lr{\frac{\Delta I_H}{I_H}}^2} \label{eq:P_H_err}
\end{align}
Ist $f$ die Motordrehzahl, so ergibt sich 
\begin{align}
	Q_2 &= W_H=  \frac{P_H}{f} = \frac{U_H\cdot I_H}{f}\label{eq:T}\\
	\Delta Q_2 &= Q_2 \cdot \sqrt{\lr{\frac{\Delta P_H}{P_H}}^2 + \lr{\frac{\Delta f}{f}}^2}.
\end{align}
Die Motorleistung ergibt sich zu
\begin{align}
	W_M &= \frac{U_M \cdot I_M}{f}\label{eq:W_M}\\
	\err{W_M}{U_M}{I_M}
\end{align}
Der Wirkungsgrad ergibt sich dann zu 
\begin{align}
\eta&=\frac{Q_{2}}{W_{M}}\\
	\err{\eta}{Q_2}{W_M}.
\end{align}

Für die Abgegebene Wärme $Q_1$ findet sich
\begin{align}
	Q_1 &= \frac{c_W \rho_W  (T_{\t{zu}}-T_{\t{ab}}) \dot{V}}{f}\label{eq:Q_1}\\
	\Delta Q_1 &= Q_1 \cdot \sqrt{ \lr{\frac{\Delta \dot V}{\dot V}}^2 +  \lr{\frac{\Delta f}{f}}^2  + \frac{\Delta T_{\t{zu}}^2 + \Delta T_{\t{ab}}^2}{(T_{\t{zu}} - T_{\t{ab}})^2}}\label{eq:Q_1_err}.\end{align}

In einem idealen System würde
\begin{align}
	|W_M |= |Q_2| + |Q_1| := W_2 \label{eq:Bil}
\end{align}
gelten. Wir werden beide Seiten der letzteren Gleichung vergleichen. Es ist
\begin{align}
\Delta W_2 =\sqrt{\Delta Q_2^2+\Delta Q_1^2}.
\end{align}


```python
#Funktionen zur späteren Verwendung. 
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
from IPython.display import Markdown, Latex, display
import pandas as pd
from scipy.optimize import curve_fit
import uncertainties
from uncertainties import ufloat as uf
from uncertainties.umath import *
from uncertainties import unumpy as unp
from uncertainties.unumpy import (nominal_values as noms,
                                  std_devs as stds)

table_counter=1

def comma_to_float(valstr):
  return float(valstr.decode("utf-8").replace(',','.'))

def round_up(n, decimals = 0):  
    multiplier = 10 ** decimals  
    return np.ceil(n * multiplier) / multiplier 

def out(name,c,unit="",zp=0,decimals=2):
    if isinstance(c, uncertainties.UFloat):
        display(Markdown("$"+name+" = " +'{:L}'.format(c)+" \   \mathrm{"+unit+"}$"))
    elif type(c) in [float,int]:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        print(c)
        
def df_out(df,title=""):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1

def usigma(g1,g2):
    return abs((g1.n-g2.n)/unp.sqrt(g1.s**2+g2.s**2))

c_w = 4180
rho_w = 997.8

#Data:
U_M = uf(26.2,0.1)
I_M = uf(2.7,0.1)
U_H = uf(5.52,0.01)
I_H = uf(1.07,0.01) * 5 #Stromwandler
V_dot = uf(294,4) * 10**(-6) / 60 #zu SI-Einheiten
T_zu = uf(21.4 + 273.15,0.1) #SI
T_ab = uf(18.5 + 273.15,0.1)
f = uf(321.9,0.2) / 60 #SI

P_H = U_H * I_H
Q_2 = P_H / f
Q_1 = c_w * rho_w * V_dot * (T_zu - T_ab) / f
W_2 = Q_1 - Q_2
W_M = U_M * I_M / f
sigma_M =usigma(W_2,W_M)
eta = Q_2 / W_M

out("P_H",P_H,"W")
out("Q_2",Q_2 ,"J")
out("W_M",W_M ,"J")
out("\eta",eta)
out("Q_1",Q_1 ,"J")
out("W_2",W_2 ,"J")
out("\sigma_{M2}",sigma_M)
```

Wir sehen also insbesondere eine signifikante Abweichung der beiden Seiten aus (25). Näheres dazu wird in der Diskussion diskutiert.

Wir wollen zunächst obenstehendes Diagramm erklären. Im Diagramm ist zu sehen, dass sich beim Betrieb als Kältemaschiene zunächst ein etwa linearer Abfall der Temperatur einstellt, der dann etwa beim Gefrierpunkt stagniert und nach einiger Zeit $t_g$, der Gefrierzeit, schließlich weiter fällt, bis sich bei etwa -20°C ein Gleichgeweicht einzustellen scheint. Die Diskontionuität beim Gefrierpunkt erklärt sich daraus, dass zum Aggregatszustandswechsel weitere Energie entzogen werden muss, damit das Wasser gefriert, die sogenannte latente Wärme. In übereinstimmung zu dieser Aussage findet sich, dass die Kühlwassertemperatur über den gesamten Messzeitraum konstant bleibt, es wird also stets die gleiche Energie pro Zeiteinheit abgeführt. Wir entnehmen nun $t_g$ dem Diagramm und berechenen daraus die Kälteleistung
\begin{align}
	P_K &= \frac{\rho_W\cdot V \cdot  \lambda_{H_2O}}{t_g}\\
	\err{P_K}{V}{t_g} 
\end{align}
wobei $ \lambda_{H_2O}$ die spezifische Schmelzwärme ist. In einem idealen System würden wir $P_K=P_H$ erwarten. Wir vergleichen die Werte.

```python
t_1=uf(225,25)
t_2=uf(385,30)
t_g=t_2-t_1
V = uf(10**(-6),10**(-7))
lam = 335 * 10**3

P_K = rho_w * V * lam / t_g
sigma_K = usigma(P_K,P_H)

out('P_K', P_K,"W")
out('\sigma_K', sigma_K)
```

Wir sehen also eine überuas signifikante Abweichung der Werte, dies wird später näher diskutiert. Die Stagnation zum Ende der Messung ist vermutlich darüber zu erklären, dass der Messaufbau nicht sonderlich stark isoliert ist und sich daher ein Gleichgewicht mit eindringender Wärme einstellt. Wird der Motor nun als Wärmepumpe betrieben, so tritt genau der umgekehrte Effekt um den Gefrierpunkt auf, jedoch wird dafür weniger Zeit benötigt, da $Q_1>>Q_2$, da die verrichtete Arbeit in Q_1 mit aufgnenommen werden kann.

### Betrieb als Wärmekraftmaschine
Wir wollen zunächst die folgenden Größen zu den Messdaten der Leerlaufmessung bestimmen:
- Elektrische Leistung $P_{el}$ analog zu oben.
- Zugeführte Wärme \begin{align}
	Q_{\t{el}} &= \frac{P_{\t{el}}}{f}\\
 \err{Q_{\t{el}}}{P_{\t{el}}}{f}
\end{align}
- Durch das Kühlwasser abgeführte Wärme $Q_{ab}$, analog zum ersten Versuchsteil.
-abgeführte Leistung
\begin{align}
	P_{\t{ab}} &= Q_{\t{ab}} \cdot f\\
	P_{\t{ab}} &= P_{\t{ab}}  \cdot \sqrt{ \lr{\frac{\Delta \dot V}{\dot V}}^2  + \frac{\Delta T_{\t{zu}}^2 + \Delta T_{\t{ab}}^2}{(T_{\t{zu}} - T_{\t{ab}})^2}}
\end{align}
- verrichtete Arbeit in einem Umlauf
\begin{align}
	P_{\t{pV}} &= W_{\t{pV}} \cdot f\\
	 \err{P_{\t{pV}}}{W_{\t{pV}}}{f}
\end{align}
wobei $W_{pV}$ und $f$ je duch mitteln aus den drei Messungen hervorgehen. 
- thermischer Wirkungsgrad
\begin{align}
	\eta_{\t{th}} &= \frac{W_{\t{pV}}}{Q_{\t{el}}}\\
	\err{\eta_{\t{th}}}{W_{\t{pV}}}{Q_{\t{el}}}
\end{align}


```python
#Leerlaufmessung:
I_l = uf(2.61,0.01) * 5 #Stromwandler
U_l = uf(11.96,0.01)
V_dot_l = uf(291,4) * 10**(-6) / 60 
T_zu_l = uf(22.9 + 273.15,0.1) 
T_ab_l = uf(18.7 + 273.15,0.1) 
f_a = np.array([276.7, 260, 256.1]) / 60 #SI
W_pV = np.array([28000, 25970, 26100]) * 10**(-4) #SI

f_l = uf(np.mean(f_a),np.std(f_a))
out('f ', f_l,"1/s")
P_el_l = U_l * I_l
out('P_{el} ', P_el_l,"W")

Q_el_l = P_el_l / f_l
out('Q_{el} ', Q_el_l, "J")

Q_ab_l = c_w * rho_w * V_dot_l * (T_zu_l - T_ab_l) / f_l
out('Q_{ab} ', Q_ab_l, "J")

P_ab_l = c_w * rho_w * V_dot_l * (T_zu_l - T_ab_l)
out('P_{ab} ', P_ab_l, "W")

W_pV_l = uf(np.mean(W_pV),np.std(W_pV))
out('W_{pV} ', W_pV_l, "J")

P_pV_l = W_pV_l * f_l
out('P_{pV}', P_pV_l, "W")

eta_th_l = W_pV_l / Q_el_l
out('\eta_{th} ', eta_th_l)
```

Nun soll die Verlustwärme
\begin{align}
	Q_V &= Q_{\t{el}} - W_{\t{pV}} -  Q_{\t{ab}} \\
 \Delta Q_V &= \sqrt{\Delta Q_{\t{el}}^2 + \Delta W_{\t{pV}}^2 + \Delta Q_{\t{ab}}^2}
\end{align}
bestimmt werden. 

```python
Q_V_l = Q_el_l - W_pV_l - Q_ab_l
out("Q_V",Q_V_l,"J")
```

Diese ist nur etwas kleiner als die durch das Kühlwasser abtransportierte Wärme und macht insgesamt etwa 38% der inverstierten Energie aus. Dies ist vermutlich auf den Aufbau zurückzuführen, der durch das Glasgehäuse zwar schön anzusehen, aber nicht sonderlich gut isoliert ist. Auch wird nicht die gesamte elektrische Leistung in Wärme umgewadelt werden, da die Heizwendel auch stark glühe und die Leitung zu dieser nicht völlig verlustfrei ist.

 \eta_{\t{eff}} &= \frac{W_D}{Q_{\t{el}}}\\

\end{align}
berechen, wobei
\begin{align}
