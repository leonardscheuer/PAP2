## Motivation
In diesem Verusch soll die Apsorption von $\alpha$-,$\gamma$- und $\beta$-Strahlung in Materie gemessen werden. Aus den Messungen sollen schlielich die (Maximal)enegien der Strahlung bestimmt werden. 

## Grundlagen
Die Eigenschaft spontan unter Energieabgabe in einen energetisch gnstigeren Zustand zu zerfallen nennt man Radioaktivitt (siehe auch vorherige Versuche). Fr die Anzahl der  verbleibenden Radionuklide gilt mit Halbwertszeit $t_H$ und anfnglichem Bestand $n_0$:
\begin{align}
	n(t) =& n_0\cdot \lr{\frac 12}^{t / t_H}\label{eq:A}\\
	=& n_0\cdot e^{-\ln(2)t/t_H}\\
	=& n_0\cdot e^{-\lambda t}\label{eq:n}
\end{align}
Die letzte Gleichkeit definiert die Zerfallskonstande $\lambda$. Wir wollen hier verschiedene Zerfallsarten betrachten:
### Alphazerfall

Im Alphazerfall wird aus dem Radionuklid $X$ ein Heliumkern emmitiert. Da der Kern nur diskrete Energiezustnde annehmen kann, wird auch beim Zerfall immer scharf derfinierte Energie frei. Instgesamt gilt:
\begin{align}
	^{A}_{Z}X \rightarrow ^{A-4}_{Z-2}X + ^4_2\alpha
\end{align}
### Betazerfall

Hier unterscheiden wir zwischen $\beta^-$- und $\beta^+$-Strahlung. 
Es wird hier Elektron/Positron und ein (Anti-)Neutriono ausgesendent:
\begin{align}
	\beta^-: & ^{A}_{Z}X \rightarrow ^{A}_{Z+1}X + ^0_{-1}e^- + ^0_0\bar{\nu}_e\\
	\beta^+: & ^{A}_{Z}X \rightarrow ^{A}_{Z-1}X + ^0_{1}e^+ + ^0_0\nu_e 
\end{align}
Die Energien sind nun nicht mehr fest verteilt, da es nun drei Prdukte gibt. Die Maximal freiwerdene Energie ist aber weiter charakteristisch. 

### Gammazerfall

Gammastrahlung ist elektromagnetische Strahlung und wird emittiert wenn ein Elektron (einen gengend Groen) Sprung in den Energiezustnden des Atoms macht. Dieses Energiespektrum ist diskret, da diese Zustnde diskret sind. 

### Absorption

Die geladenen Teilchen der Alpha- und Betastrahlung werden in Materie grtenteils durch Stoe und Wechselwirkungen mit dieser gebremst. Die Energieabgabe ist dabei etwa kontinuierlich und invers proportional zur Geschwindigkeit. Ein Teilchen mit definierter Energie hat also immer etwa gleiche Reichweite. Fr Alphateilchen ist die Energie also mit der dicke des Materials zum erreichen der kritischen Absorberdicke, bei welcher abrupt fast keine Teilchen mehr durchkommen, bestimmbar. Da Betateilchen (Elektronen) keine feste Anfangsenergie besitzen und strker vom Material abgelenkt werden, ist hier bestenfalls ein verwaschenes Spektrum bestimmbar. 
Fr Gammateilchen gilt nherungsweise das Lambert-Beer-Gesetz:
\begin{align}
	n(t) = n_0 \cdot e^{-\mu x} \label{eq:lb}
\end{align}
wobei $\mu$ ein materialabhngiger Schwchungskoeffizient ist. Er ergibt sich vorrangig aus den folgenden Wechselwirkungen:
- Photoeffekt
- Comptonstreuung: Inelastischer Sto mit einem Elektron in der Atomhlle, wodurch dieses gestreut wird.
- Paarbildung: Bei entsprechend hoher Energie kann sich das Gammaquant aufteilen in ein Teilchen-Antiteilchen-Paar. Dies passiert wahrscheinlicher in der Nhe eines Atomkerns, der den Recoil (berschssiger Impuls) aufnehmen kann.

### Absorption von Betastrahlung

Wir tragen die Zhlrate bro Sekunde abzglich des Untergrundes ber die kumulative Dicke der Aluminiumplatten auf. Fr die Zhlrate pro Sekunde ergibt sich mit $N$ gemessenen Ereignissen pro Zeit $t$ (bzw. $N_0$,$t_0$ fr den Untergrund):
\begin{align}
	n - n_0^\beta &= \frac{N }{t}- \frac{N_0^\beta}{t_0^\beta}\\
	 \Delta (n - n_0^\beta) &= \sqrt{\lr{\frac{\Delta N}{t}}^2 + \lr{\frac{\Delta N_0^\beta}{t_0^\beta}}^2}
\end{align}
Wir fitten eine Funktion der Form
\begin{align}
	n(x) = a \cdot e^{-b\cdot x} - c
\end{align}
an dessen Nullstelle $x_{crit}$ lsst sich die Maximale Energie der Betastrahlung berechnen, dafr errechnet man die Flchendichte
\begin{align}
	R^\beta &= R_{ES}^\beta + \rho \cdot x_\t{crit}\\
	\Delta R^\beta &=  \rho \cdot \Delta x_\t{crit}
\end{align}
Wobei $R_{ES}=0.130 g/cm^2$ die durch den Prperataufbau vorhandene Flchendichte ist und $\rho=2.6989g/cm^3$ die Dichte von Aluminium. Aus der nachfolgenden Abb. knnen wir dann die Energie entnehmen:


Abb.2: Reichweite von -Strahlung in Aluminium und -Strahlung in Luft (Quelle: Script)




```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit
from scipy.stats import chi2

def comma_to_float(valstr):
    return float(valstr.decode("utf-8").replace(',','.'))

def expon(x, a, b, c):
    return a * np.exp(-b * x) - c

t1, x1, N1, N1_err = np.genfromtxt("v2.csv", delimiter = ';', skip_header = 1, converters = {0:comma_to_float, 1:comma_to_float, 2:comma_to_float, 3:comma_to_float}, unpack = True)

n1 = N1 / t1
n1_err = N1_err / t1

n1 = n1[:-3-1] - n1[-1]
n1_err = np.sqrt(n1_err[:-3-1]**2 + n1_err[-1]**2)

x1 = x1[:-3-1]

plt.errorbar(x1, n1, yerr = n1_err, label = 'Messdaten ohne Untergrund', capsize = 3, fmt = 'bx')
plt.yscale('log')
plt.xlabel('Absorberdicke / mm')
plt.ylabel('Zhlrate minus Untergrund / 1/s')
plt.title('Absorption von Betastrahlung')
plt.ylim(bottom = 10**(-2))

popt1, pcov1 = curve_fit(expon, x1, n1, sigma = n1_err)

s = np.linspace(x1[0], x1[-1] + 0.55, 100)

crit_min = np.array([popt1[0] - np.sqrt(pcov1[0][0]), popt1[1] + np.sqrt(pcov1[1][1]), popt1[2] + np.sqrt(pcov1[2][2])])
crit_max = np.array([popt1[0] + np.sqrt(pcov1[0][0]), popt1[1] - np.sqrt(pcov1[1][1]), popt1[2] - np.sqrt(pcov1[2][2])])

plt.plot(s, expon(s, *popt1), label = 'Fit von n(x) = $a\cdot e^{-b \cdot x}-c$')
plt.plot(s, expon(s, *crit_min), color = 'red', label = 'Fehlerkurve')
plt.plot(s, expon(s, *crit_max), color = 'red')
plt.legend()
plt.text(0.2, 0.09, 'a = ({0:.1f}  {1:.1f})/s\nb = ({2:.2f}  {3:.2f})/mm\nc = ({4:.2f}  {5:.2f})/s'.format(popt1[0], np.sqrt(pcov1[0][0]), popt1[1], np.sqrt(pcov1[1][1]), popt1[2], np.sqrt(pcov1[2][2])))
plt.grid("dotted")
```

```python
chi2_ = np.sum((expon(x1, *popt1) - n1)**2/ n1_err**2)
dof = len(n1) - 3
chi2_red = chi2_ / dof

print("chi^2 = {0:.1f}".format(chi2_))
print("chi^2_red = {0:.1f}".format(chi2_red))
```

```python
prob = round(1 - chi2.cdf(chi2_, dof), 2) * 100
print("Fitwahrscheinlichkeit: {0}%".format(prob))
```

```python
x_crit1 = np.log(popt1[0] / popt1[2]) / popt1[1]
x_crit1_err = np.max(np.array([np.abs(np.log(crit_min[0] / crit_min[2]) / crit_min[1] - x_crit1), np.abs(np.log(crit_max[0] / crit_max[2]) / crit_max[1] - x_crit1)]))

print('Maximalreichweite: ({0:.1f}  {1:.1f})mm'.format(x_crit1, x_crit1_err))
```

```python
rho_Al = 2.6989
R_ES = 0.130

print('Flchendichte: ({0:.2f}  {1:.2f})g/cm^2'.format(R_ES + rho_Al * x_crit1 / 10, rho_Al * x_crit1_err / 10))
```

### Absorption von $\gamma$-Strahlung
Analog zu oben wird die Zhlrate ber der Absorberdicke aufgetragen. Der Schwchungskoeffizient durch Fit bestimmt. Mit der Dichte $\rho_{Pb}$ von Blei wird dann der Massenschwchungskoeffizient bestimmt:
\begin{align}
	k &= \frac \mu {\rho_{Pb}}\\
	\Delta k &= \frac {\Delta \mu}{\rho_{Pb}}
\end{align}

```python
def lambert_beer(x, n_0, mu):
    return n_0 * np.exp(- mu * x)

n0 = 91 / (5 * 60) #in 1/s
n0_err = np.sqrt(n0)

t2, x2, N2, N2_err = np.genfromtxt("v3.csv", delimiter = ';', skip_header = 1, converters = {0:comma_to_float, 1:comma_to_float, 2:comma_to_float, 3:comma_to_float}, unpack = True)

n2 = N2 / t2
n2_err = N2_err / t2

n2 = n2 - n0
n2_err = np.sqrt(n2_err**2 + n0_err**2)

plt.errorbar(x2, n2, yerr = n2_err, label = 'Messdaten ohne Untergrund', capsize = 3, fmt = 'bx')
plt.yscale('log')
plt.xlabel('Absorberdicke / mm')
plt.ylabel('Zhlrate ohne Untergrund / 1/s')
plt.title('Absorption von Gammastrahlung')

popt2, pcov2 = curve_fit(lambert_beer, x2, n2, sigma = n2_err, p0 = (38.7, 0.063))

s = np.linspace(x2[0], x2[-1], 100)

plt.plot(s, lambert_beer(s, *popt2), label = 'Fit von n(x) = $n_0\cdot e^{-\mu \cdot x}$')
plt.legend()
plt.text(2, 5, 'n_0 = ({0:.1f}  {1:.1f})/s\nmu = ({2:.4f}  {3:.4f})/mm'.format(popt2[0], np.sqrt(pcov2[0][0]), popt2[1], np.sqrt(pcov2[1][1])))
plt.grid('dotted')
```

```python
chi2_ = np.sum((lambert_beer(x2, *popt2) - n2)**2/ n2_err**2)
dof = len(n2) - 2
chi2_red = chi2_ / dof

print("chi^2 = {0:.1f}".format(chi2_))
print("chi^2_red = {0:.1f}".format(chi2_red))
```

```python
prob = round(1 - chi2.cdf(chi2_, dof), 2) * 100
print("Fitwahrscheinlichkeit: {0}%".format(prob))
```

```python
rho_Pb = 11.342

k = popt2[1] / rho_Pb * 10
k_err = np.sqrt(pcov2[1][1]) / rho_Pb * 10

print('Massenchwchungskoeffizient: ({0:.4f}  {1:.4f})cm^2/g'.format(k, k_err))
```

### Bestimmung der Aktivitt
Nun soll die Aktivitt des Gammastrahlers berechnet werden. Die Aktivitt $n$ auf der gemessenen Flche, ergibt hochgerechnet die Gesamtaktivitt

\begin{align}
	A = n\cdot 4 \frac{d^2}{r^2}
\end{align}

Wobei $d$ der Abstand zum Detektor und $r$ der Detektorradius ist.

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from matplotlib.pyplot import figure
import numpy as np
from IPython.display import Markdown, Latex, display
import pandas as pd
from scipy.optimize import curve_fit
import uncertainties
from uncertainties import ufloat as uf
from uncertainties.umath import *
from uncertainties import unumpy as unp
from uncertainties.unumpy import (nominal_values as noms,
                                  std_devs as stds)
from scipy.signal import *

table_counter=1

diag_dpi=100

def round_up(n, decimals = 0):  
    multiplier = 10 ** decimals  
    return np.ceil(n * multiplier) / multiplier 

def out(name,c,unit="",zp=0,decimals=2):
    if isinstance(c, uncertainties.UFloat):
        display(Markdown("$"+name+" = " +'({:L})'.format(c)+" \   \mathrm{"+unit+"}$"))
    elif type(c) in [float,int]:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        print(c)
        
def df_out(df,title=""):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1

def usigma(g1,g2):
    return abs((g1.n-g2.n)/unp.sqrt(g1.s**2+g2.s**2))

def outwstd(name,c,Dc,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = (" +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+") "+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))

def out_sys_std(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        pass
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c[0]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+ "{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals) +" ) " + unit + " $"))
    else:
        display(Markdown( "$ "+name+" = ( " + "{:.{}f}".format(round(c[0]*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))
        
def out_no_error(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \cdot 10^{ "+str(zp)+" }  " + unit + " $"))

def df_out(df,title=""):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1
    
def sigma(g1,Dg1,g2,Dg2):
    return abs((g1-g2)/np.sqrt(Dg1**2+Dg2**2))


A_lit = 3.7 * 0.5**(7/5.272)

r = 1.4 / 2
r_err = 0.1 / 2

l = 4.0
l_err = 0.1

eps = 0.04
eps_err = 0.01

d, d_err, n3, n3_err = np.genfromtxt("v4.csv", delimiter = ';', skip_header = 1, converters = {0:comma_to_float, 1:comma_to_float, 2:comma_to_float, 3:comma_to_float}, unpack = True)

n3 = (n3 / 60 - n0) * 10**(-6) #in MBq
n3_err = np.sqrt((n3_err / 60)**2 + n0_err**2) * 10**(-6)

A = 4 * n3 * d**2 / r**2 / eps
A_err = A * np.sqrt((2 * d_err/d)**2 + (2 * r_err/r)**2 + (eps_err/eps)**2 + (n3_err / n3)**2)


df=pd.DataFrame(data={"$d$ [cm]":d,"$A$ [kBq]":A,"$\Delta A$ [kBq]":A_err,"$\sigma_{A,lit}$":np.abs(A - A_lit)/A_err})

df_out(df,"Aktivitt")
```

Die obige Nherung funktioniert am besten fr groe Verhtnisse zwischen Detektor-Prperatabstand und Zhlrohrdurchmesser. Wir wollen nun, da der Detektor Ausdehnung besitzt, eine Raumwinkelkorrektur mit einem Korrekurfaktor 
\begin{align}
k_1 &=  \frac{(d + \frac l2)^2}{d^2} =(1 + \frac l{2d})^2\\
\Delta k_1 &= 2(1 + \frac l{2d})\cdot \frac l{2d}\sqrt{\lr{\frac{\Delta d}{d}}^2 + \lr{\frac{\Delta l}{l}}^2}
\end{align}
Aus der Nachfolgenen Abbildung ist dies Einsichtig:


Abb. 5: Extremflle des Raumwinkels (Quelle: Script)

Im Korrekturfaktor gleichen unter- und berschtzte Bereiche gerade aus. Wir nennen die neu erhaltene Aktivitt $A_1$.


```python
k1 = (1 + l / (2 * d))**2
k1_err = (1 + l / (2 * d)) * l / d * np.sqrt((l_err / l)**2 + (d_err / d)**2)

A1 = k1 * A
A1_err = A1 * np.sqrt(4 * (d_err**2 + l_err**2 / 2) / (d + l / 2)**2 + (2 * r_err/r)**2 + (eps_err/eps)**2 + (n3_err / n3)**2)


df["$k_1$"]=k1
df["$\Delta k_1$"]=k1_err
df["$A_1$ [kBq]"]=A1
df["$\Delta A_1$ [kBq]"]=A1_err
df["$\sigma_{A_1,lit}$"]=np.abs(A1 - A_lit)/A1_err

df_out(df,"Aktivitt mit Raumwinkelkorrektur")
```

Nun soll auch die Absorption in der Prperatskapsel miteinbezogen werden. Dazu wird der Massenschwchungskoeffizient aus dem ersten Teil genutzt. Sei
\begin{align}
	\mu &= k \cdot \rho\\
	\Delta \mu &= \rho \cdot \Delta k
\end{align}
wobei $\rho=7,9 g/cm^3$ die Dichte der Prperatkapsel ist. Daraus ergibt sich mit der Dicke der Prperatkapsel $x=(1,4\pm0,1)mm$ der Korrekturfaktor
\begin{align}
	k_2 &= e^{\mu x} \\
	 \Delta k_2 &= k_2\cdot \sqrt{(\mu \Delta x)^2 + (x\Delta \mu)^2}
\end{align}
Wir nennen die nun zweifach korrigierte Aktivitt $A_2$

```python
rho_PK = 7.9

x = 1.4
x_err = 0.1

mu = k * rho_PK
mu_err = rho_PK * k_err
out("\mu",uf(mu, mu_err))
```

```python
k2 = np.exp(k * x)
k2_err = k2 * np.sqrt((x * k_err)**2 + (k * x_err)**2)

A2 = k2 * A1
A2_err = A2 * np.sqrt((k2_err / k2)**2 + (A1_err / A1)**2)

k_2=uf(k2,k2_err)
out("k_2",k_2)
df["$A_2$ [kBq]"]=A2
df["$\Delta A_2$ [kBq]"]=A2_err
df["$\sigma_{A_2,lit}$"]=np.abs(A2 - A_lit)/A2_err

df_out(df.round(3),"Aktivitt mit Raumwinkel- und Absorptionskorrektur")
```

### Absorption von $\alpha$-Strahlung
Wir wollen die Energie der Alphastrahlung berechnen, dazu berechenen wir zunchst die Reichweite $s_1$. Dazu berechen wir zunchst beim Abfall mit linearem Fit den kritischen Druck, auf Hlfte der Gerade, welcher dann, mit $s_0$ Abstand zwischen Zhlrohr und Prperat, ergibt:
\begin{align}
	s_1 &= \frac{p}{p_0}s_0\\
	 \err{s_1}{p}{s_0}\\
\end{align}


```python
def line(x, m, b):
    return m * x + b

p, p_err, n4, n4_err = np.genfromtxt("v5.csv", delimiter = ';', skip_header = 1, converters = {0:comma_to_float, 1:comma_to_float, 2:comma_to_float, 3:comma_to_float}, unpack = True)

n3 = n3 / 60 #in Bq
n3_err = n3_err / 60

plt.errorbar(p, n4, xerr = p_err, yerr = n4_err, label = 'Messdaten', capsize = 3, fmt = 'bx')
plt.xlabel('Druck / mbar')
plt.ylabel('Zhlrate / 1/s')
plt.title('Absorption von Alphastrahlung')
plt.xlim(right = 1000)

popt4, pcov4 = curve_fit(line, p[6:12], n4[6:12], sigma = n4_err[6:12])
plt.plot(p[5:13], line(p[5:13], *popt4), label = 'Fit: n(p) = ({0:.0f}/mbar $\cdot$ p + {1:.0f})/s'.format(popt4[0], popt4[1]))
plt.legend()
plt.grid("dotted")
```

```python
n_12 = 6540
n_12_err = 60

p_12 = (n_12 - popt4[1]) / popt4[0]
p_12_err = p_12 * np.sqrt(pcov4[0][0] / popt4[0]**2 + (n_12_err**2 + pcov4[1][1]) / (n_12 - popt4[1])**2)

print('Druck bei halber Zhlrate: ({0:.0f}  {1:.0f})mbar'.format(p_12, p_12_err))
```

```python
s_0 = 4.2
s_0_err = 0.1

p0 = 1013

s_1 = p_12 * s_0 / p0
s_1_err = s_1 * np.sqrt((p_12_err / p_12)**2 + (s_0_err / s_0)**2)

out("s_1",uf(s_1,s_1_err),"cm")
```

Nun soll noch die Absorption durch das Zhlrohrfenster (Glimmer) korrigiert werden, analog zu oben:

```python
rho_gl = 2.25
rho_gl_err = 0.01

s_2 = s_1 + rho_gl / 1.43
s_2_err = np.sqrt(s_1_err**2 + (rho_gl_err / 1.43)**2)

out("s_2",uf(s_2,s_2_err),"cm")
```

Die Schutzschicht um das Prperat besitzt eine quivalente Luftdicke von 0.68 cm, dies wird hinzuaddiert.

```python
s = s_2 + 0.68
s_err = s_2_err

print('s = ({0:.2f}  {1:.2f})cm'.format(s, s_err))
out("s",uf(s,s_err),"cm")
```
