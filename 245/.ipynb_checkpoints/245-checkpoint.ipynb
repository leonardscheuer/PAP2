## Grundlagen

### Induktionsgesetz
Wir betrachten eine Leiterschleife mit $N$ Windungen und eingeschlossener Flche $A$. Ist $\phi$ der Magnetische  Fluss durch die Flche, so gilt (dies lsst sich sofort aus der ensprechenden Maxwellgleichung sehen) mit der Definition der Spannung
\begin{align}
U=\int \vec{E} d\vec{s}
\end{align}
das Induktionsgesetz:
\begin{equation}
U(t)=-\frac{d}{d t} \phi=-B A N \omega \sin (\omega t)
\end{equation}
Wobei sich die letzte Gleichheit ergibt, wenn die Flachspule um eine in ihrer Flche liegende Achse $\vec{\omega}$ rotiert wird. $B$ ist die magnetische Flussdichte, die hier zunchst als homogen angenommen wird. Ist das B-Feld periodisch mit $\Omega$ sinusidal richtungswechselnd, so ergibt sich mit $B$ als Amplitude des schwingenden Feldes:
\begin{align}
U=BAN \Omega \cos(\varphi)\sin(\Omega )
\end{align}
wobei $\varphi$ der Neigungswinkel zwischen der Flchennormale der Flachspule und dem B-Feld ist.

### Helmholtzspulen
Eine Helmholtzspule ist eine Anordnung von zwei Spulen, deren Abstand zueinander gerade der Radius ist. Dieser Abstand ist gerade so gewhlt, dass auch die zweite Ableitung des Feldes im Zentrum verschwindet (erste verschwindet wegen Symmetrie). Das Feld weit dort also eine hohe Homogenitt auf. Mittels Biot-Savart bestimmt sich das Feld so zu:
\begin{align}
B = {\left ( \frac{4}{5} \right )}^{3/2} \frac{\mu_0 n I}{R}
\end{align}
Wobei $n$ hier die Anzahl der Windungen der Spulen ist.

Die Erdoberflche durchdringen die Magentfeldlinien mit einem Positionsabhngigen Inkanationswinkel $\alpha$. $\vec{B}$ knnen wir in Horizontal- und Vertikalkomponente aufspalten. ## Versuche
### 1 Vorversuch
Qualitativ wird ein Stabmagnet in verschiedener Geschwindigkeit durch eine Spule bewegt, um Induktion zu beobachten.
### 2 Induktionsgesetz
Hier wird die Induktionsspannung (Spitze-Spitze-Spannung) in einer rotierenden Induktionsspule (mit dem Osszilloskop) gemessen:
- Bei konstantem (Helmholtz-)Spulenstrom von etwa 4A fr Drehfrequenzen zwischen 3Hz bis zu 15Hz in Schritten von 3Hz.
- Bei konstanter Drehfrequenz von etwa 10Hz bei vernderlichem Spulenstrom zwischen 0.5A bis 4.5A in Schritten von 0.5A

### 3 Induktionsspannung bei periodischem Feldstrom (Lufttransformator)
Nun wird die Helmholtzspule an den Leistungsfunktionengenerator angeschlossen. Es wird die Induktionsspannung gemessen:
- Bei konstantem Wechselstrom 100Hz in verschiedenen Winkeln $\varphi$ in Schritten von 30. 
- Im Frequenzbereich zwischen 20Hz und 2kHz. In Schritten von 20Hz im Bereich unter 200Hz, darber 200Hz. Zustzlich werden Strom und Spannug der Helmholtzspulen aufgenommen. 
Schlielich berzeugt man sich qualitativ davon, dass bei Drehung der Sekundrspule eine Schwebung mit gleicher Frequenz der Induktionsspannug entsteht. 

### 4 Bestimmung des Erdmagnetfeldes durch Kompensation
Der Aufbau wird grob in Nord-Sd-Richtung ausgerichtet. Es wird gemessen:
- die Induktionsspannung bei einer Rotation der Spule von etwa 15Hz ohne knstliches Magnetfeld.
- der Helmholtzspulenstrom, bei welchem die Induktionsspannung minimal wird. Dort wird auch die verbleibende Induktionsspannung und Drehfrequenz (etwa 15Hz whlen) gemessen.

## Auswertung
### Induktionsgesetz
Wir tragen zunchst die Induktionsspannung (halbe Spitze-Spitze-Spannung) ber Frequenz und Spulenstrom auf. ber ersteres bestimmen wir durch linearen Fit das Magnetfeld B. Die Steigung der Geraden ist nach obiger Formel gerade $m=2 \pi BAN$, also 
\begin{align}
B_{i}&=m/ 2 \pi NA\\
\Delta B_{i} &= \Delta m  / 2 \pi NA
\end{align}
Wir wollen den Wert Vergleichen mit demjenigem $B_H$, welches erhalten wird, wenn Gl. (4) verwendet wird.

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from matplotlib.pyplot import figure
import numpy as np
from IPython.display import Markdown, Latex, display
import pandas as pd
from scipy.optimize import curve_fit
import uncertainties
from uncertainties import ufloat as uf
from uncertainties.umath import *
from uncertainties import unumpy as unp
from uncertainties.unumpy import (nominal_values as noms,
                                  std_devs as stds)
from scipy.signal import *

table_counter=1

diag_dpi=100

def round_up(n, decimals = 0):  
    multiplier = 10 ** decimals  
    return np.ceil(n * multiplier) / multiplier 

def out(name,c,unit="",zp=0,decimals=2):
    if isinstance(c, uncertainties.UFloat):
        display(Markdown("$"+name+" = " +'({:L})'.format(c)+" \   \mathrm{"+unit+"}$"))
    elif type(c) in [float,int]:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        print(c)
        
def df_out(df,title=""):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1

def usigma(g1,g2):
    return abs((g1.n-g2.n)/unp.sqrt(g1.s**2+g2.s**2))

def outwstd(name,c,Dc,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = (" +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+") "+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))

def out_sys_std(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        pass
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c[0]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+ "{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals) +" ) " + unit + " $"))
    else:
        display(Markdown( "$ "+name+" = ( " + "{:.{}f}".format(round(c[0]*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))
        
def out_no_error(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \cdot 10^{ "+str(zp)+" }  " + unit + " $"))

def df_out(df,title=""):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1
    
def sigma(g1,Dg1,g2,Dg2):
    return abs((g1-g2)/np.sqrt(Dg1**2+Dg2**2))

def quad_add(arr):
    quad_sum=0
    for i in arr:
        quad_sum+=i**2
    return (quad_sum**0.5)/len(arr)

def fit_line(x, m, n): 
    return m * x + n 
```

```python
Df_U_to_f=pd.read_csv('1_1.csv')  #Read Data
Df_U_to_f["U [V]"]=0.5* Df_U_to_f["U [V]"]
Df_U_to_f["U [V]"]=0.5* Df_U_to_f["U [V]"]
popt_f, pcov_f = curve_fit(fit_line, Df_U_to_f["f [Hz]"], Df_U_to_f["U [V]"], sigma = np.sqrt(quad_add([Df_U_to_f["U [V]"],Df_U_to_f["f [Hz]"]])))

x = np.linspace(3, 15.2, 2) 

fig, (plt1, plt2) = plt.subplots(1, 2, sharey=True)
fig.set_size_inches(15,6)
fig.set_dpi(diag_dpi)

plt1.plot(x, fit_line(x, *popt_f),label="Ausgleichsgerade") 
plt1.grid(ls='dotted')
plt1.errorbar(Df_U_to_f["f [Hz]"], Df_U_to_f["U [V]"], yerr = Df_U_to_f["U [V]"], xerr = Df_U_to_f["f [Hz]"], fmt = 'bx', capsize = 3)
 
plt1.set_title('Diag. 1: Induktionsspannung zu f[Hz]\n (Spulenstrom $4,0 \pm 0,1$A)') 
plt1.set(xlabel='f[Hz]', ylabel='$U_i$[V]')
_=plt1.legend() 

Df_U_to_I=pd.read_csv('1_2.csv')  #Read Data
Df_U_to_I["U [V]"]=0.5* Df_U_to_I["U [V]"]
Df_U_to_I["U [V]"]=0.5* Df_U_to_I["U [V]"]
popt_I, pcov_I = curve_fit(fit_line, Df_U_to_I["I [A]"], Df_U_to_I["U [V]"], sigma = np.sqrt(quad_add([Df_U_to_I["U [V]"],Df_U_to_I["I [A]"]])))

x = np.linspace(0.1, 4.5, 2) 

plt2.plot(x, fit_line(x, *popt_I),label="Ausgleichsgerade") 
plt2.grid(ls='dotted')
plt2.errorbar(Df_U_to_I["I [A]"], Df_U_to_I["U [V]"], yerr = Df_U_to_I["U [V]"], xerr = Df_U_to_I["I [A]"], fmt = 'bx', capsize = 3)
 
plt2.set_title('Diag. 2: Induktionsspannung zu I[A]\n (Drehfrequenz $(10,00 \pm 0,04)$Hz)') 
plt2.set(xlabel='I[A]')
_=plt2.legend() 

N=4000
A=0.00417
n=124
r=0.147
mu_0=1.25663706212*10**(-6)

I=uf(4,0.1)

B_i=uf(popt_f[0]/(2*np.pi*N*A),np.sqrt(pcov_f[0,0])/(2*np.pi*N*A))
B_H=I*8*mu_0*n/(np.sqrt(125)*r)
out("B_i",B_i*10**3,"mT")
out("B_H",B_H*10**3,"mT")
out("\\sigma_{B_i, B_H}",usigma(B_H,B_i),"\\sigma")
```

Die Ergebnisse unterscheiden sich signifikant.  Eine Erklrung hierfr ist dass der Raum zwischen dern Helmholtzspulen nicht, wie im idealisierten Fall, frei, sondern gefllt durch die Aufhngung der Sekundrspule. Eine weitere Betrachtung wird in der Diskussion erfolgen.

### Induktionsspannung bei periodischem Feldstrom
Wir geben zunchst die Messwerte der Induktionsspannung zum Winkel an. Jedoch ist nur der Betrag der Spannung gemessen worden, der negative Cosinus-Faktor aus (3) ist daher nur im Betrag sichtbar. Rechnet man den Vorzeichenwechsel mit ein, so erhlt man das zweite Diagramm und erhlt wie erwartet einen Cosinus.


```python
Df_U_to_f=pd.read_csv('2_1.csv')  #Read Data
Df_U_to_f["U [V]"]=0.5* Df_U_to_f["U [V]"]
Df_U_to_f["U [V]"]=0.5* Df_U_to_f["U [V]"]

x = np.linspace(3, 15.2, 2) 

fig, (plt1,plt2) = plt.subplots(1, 2)
fig.set_size_inches(15,6)
fig.set_dpi(diag_dpi)

plt1.grid(ls='dotted')
plt1.errorbar(Df_U_to_f[" []"], Df_U_to_f["U [V]"], yerr = Df_U_to_f["U [V]"], xerr = Df_U_to_f[" []"], fmt = 'bx', capsize = 3)
plt1.set_title('ohne Vorzeichenkorrektur') 

fig.suptitle('Diag. 3: Induktionsspannung zu  []\n (Wechselspannung ($103,0 \pm 0,1$)Hz)') 
_=plt1.set(xlabel=' []', ylabel='$U_i$[V]')

Df_U_to_f["U [V]"]=Df_U_to_f["U [V]"]*(-1)**(Df_U_to_f[" []"].between(90, 180))
plt2.set_title('mit Vorzeichenkorrektur') 

def amp_cos(x,amp):
    return amp*np.cos(x*(2*np.pi/360))

def amp_shift_cos(x,amp,shift):
    return amp*np.cos((x+shift)*(2*np.pi/360))


x = np.linspace(0, 180, 500) 

popt_a, pcov_a = curve_fit(amp_cos, Df_U_to_f[" []"], Df_U_to_f["U [V]"], sigma = np.sqrt(quad_add([Df_U_to_f["U [V]"],Df_U_to_f[" []"]])))

plt2.plot(x, amp_cos(x, *popt_a),label="Cosinus mit angepasster Amplitude") 

popt_a, pcov_a = curve_fit(amp_shift_cos, Df_U_to_f[" []"], Df_U_to_f["U [V]"], sigma = np.sqrt(quad_add([Df_U_to_f["U [V]"],Df_U_to_f[" []"]])))

plt2.plot(x, amp_shift_cos(x, *popt_a),label="Cosinus mit angepasster Amplitude und $ \\alpha $-shift") 

plt2.grid(ls='dotted')
plt2.errorbar(Df_U_to_f[" []"], Df_U_to_f["U [V]"], yerr = Df_U_to_f["U [V]"], xerr = Df_U_to_f[" []"], fmt = 'bx', capsize = 3)
 
_=plt2.set(xlabel=' []', ylabel='$U_i$[V]')
_=plt2.legend() 
```

Wir stellen jetzt das Verhltnis $U/V$ aus Induzierter zu angelegter Spannung dar.

```python
Df_2_2=pd.read_csv('2_2.csv')  #Read Data
Df_2_2["U [V]"]=0.5* Df_2_2["U [V]"]
Df_2_2["U [V]"]=0.5* Df_2_2["U [V]"]
Df_2_2["V [V]"]=0.5* Df_2_2["V [V]"]
Df_2_2["V [V]"]=0.5* Df_2_2["V [V]"]
Df_2_2["U/V"]=Df_2_2["U [V]"]/Df_2_2["V [V]"]
Df_2_2["U/V"]=np.sqrt((Df_2_2["U [V]"]/Df_2_2["V [V]"])**2+(Df_2_2["V [V]"]/Df_2_2["V [V]"]**2)**2)

plt.errorbar(Df_2_2["f [Hz]"], Df_2_2["U [V]"], yerr = Df_2_2["U/V"], xerr = Df_2_2["f [Hz]"], fmt = 'bx', capsize = 3)

plt.title('Diag. 4: U/V abhngig von der Frequenz (Lufttransformator)') 
plt.ylabel('U/V') 
plt.xlabel('f[Hz]') 

plt.grid(ls='dotted')
```

Wir sehen, dass sich in kleinen Frequenzbereichen ein Starker Anstieg befindet, dann nur noch ein langsamer. In kleinen Frequenzbereichen dominiert der ohmsche Widerstand, in hheren die Impendanz der Spule. Der (Induktive) Widerstand der Spule ergibt sich dann zu
\begin{align}
Z=\frac{U_{eff}}{I_{eff}}=\frac{\hat{U}}{\hat{I}}=R_0+2\pi f L i
\end{align}
wobei $R_0$ der ohmsche Widerstand und $L$ die Induktivitt der Spule ist. Wir sind am Betrag
\begin{align}
|Z|=\sqrt{R_0^2+(2\pi f L)^2} 
\end{align}
interessiert. Wir fitten diese Funktion an um passende $R_0$ und $L$ zu erhalten.

```python
Df_2_2["R []"]=Df_2_2["V [V]"]/Df_2_2["I [A]"]
Df_2_2["R []"]=np.sqrt((Df_2_2["V [V]"]/Df_2_2["I [A]"])**2+(Df_2_2["I [A]"]/Df_2_2["I [A]"]**2)**2)

popt_R, pcov_R = curve_fit(fit_line, Df_2_2["f [Hz]"], Df_2_2["R []"], sigma = np.sqrt(quad_add([Df_2_2["f [Hz]"],Df_2_2["R []"]])))

x = np.linspace(0, 2000, 2) 

def abs_z(x,R,L):
    return np.sqrt(R**2+(2*np.pi*x*L)**2)

popt_Z, pcov_Z = curve_fit(abs_z, Df_2_2["f [Hz]"], Df_2_2["R []"], sigma = np.sqrt(quad_add([Df_2_2["f [Hz]"],Df_2_2["R []"]])))
R_0=uf(popt_Z[0],np.sqrt(pcov_Z[0,0]))
L=uf(popt_Z[1],np.sqrt(pcov_Z[1,1]))

fig, (plt1) = plt.subplots(1, 1)
fig.suptitle('Diag. 5: Widerstand (U/I) in Abhngigkeit der Frequenz') 
fig.set_size_inches(15,6)
fig.set_dpi(diag_dpi)
plt1.grid(ls='dotted')
plt1.plot(x, fit_line(x, *popt_R),label="Ausgleichsgerade") 
plt1.plot(x, abs_z(x, *popt_Z),label="gefittetes |Z|") 

plt1.errorbar(Df_2_2["f [Hz]"], Df_2_2["R []"], yerr = Df_2_2["R []"], xerr = Df_2_2["f [Hz]"], fmt = '.b',markersize=4, capsize = 0, ecolor="r",  elinewidth=2, label="Messwerte")
_=plt1.set(xlabel='f [Hz]', ylabel='R []')
_=plt1.legend() 

out("L",L*10**3,"mH")
out("R_0",R_0,"\Omega")
```

Wir sehen, dass sich die Kurve sehr linear verhlt, der Einfluss des Ohmschen Widerstandes ist also insbesondere fr grere Frequenzen zu vernachlssigen. Der gemessene Strom sinkt mit steigender Frequenz ab, da die Impendanz steigt.

Wir vergleichen mit der Theoretischen Formel ([Quelle:Wikipedia](https://de.wikipedia.org/wiki/Helmholtz-Spule#induktivit%C3%A4t)) :
\begin{align}
L=2\cdot n^2\cdot R\cdot \mu _0 \cdot \left({\pi R \over {R+2R/2{,}2}}+{4{,}941\over 4\pi}\right)
\end{align}

```python
L_theo=2*(n**2)*r*mu_0*(np.pi*r/(r+2*r/2.2) + 4.941/4*np.pi)
out("L",L_theo*10**3,"mH")
out("\\sigma_{L, theo}",usigma(L,uf(L_theo,0)),"\\sigma")
```

### Bestimmung des Erdmagnetfeldes
Wir stellen Gl. (2) um und erhalten:
\begin{align}
B_{E}&=\frac{U_i}{2 AN \pi f} =\frac{U_{iSS}}{4 AN \pi f}\\
\Delta B_{E}&= B_E \sqrt{\left( \frac{\Delta U_{iSS}}{U_{iSS}} \right) ^2+\left( \frac{\Delta f}{f} \right) ^2}
\end{align}
wobei $U_{iSS}$ die Spitze-Spitze-Spannung ist. Wir erhalten die Strke des Erdmagnetfeldes:

```python
U_SS=uf(142e-3,2e-3)
f = uf(14.8,0.3)
B_E=U_SS/(4*A*N*np.pi*f)
out("B_E",B_E*10**6,"\\mu T")
```

Wir bestimmen jetzt die Vertikalkomponente aus dem gemessenen Kompensationsstrom (entspr. Gl. (4) und erstem Teil der Auswertung), die Horizontalkomponente aus der verbleibenden Induktionsspannung am Minimum und den Inklanationswinkel aus ersterem:

```python
I_komp=uf(52.95e-3,0.02e-3)
U_min=uf(50e-3,3e-3)


B_vert=(I_komp*8*mu_0*n)/(np.sqrt(125)*r)
B_hor=U_min/(4*A*N*np.pi*f)


out("B_{E,V}",B_vert*10**6,"\\mu T")
out("B_{E,H}",B_hor*10**6,"\\mu T")

a=unp.arcsin(B_vert/B_E)*(360/(2*np.pi))

out("\\alpha",a,"")
```

Wir wollen dies nun mit den Literaturwerten vergleichen (Abgerufen 3.3.22, Unsicherheit an Schwankungen absgeschtzt, [Ptp](https://www.ptb.de/cms/nc/ptb/fachabteilungen/abt2/fb-25/ag-251/live-daten-erdmagnetfeldmessung.html)):

```python
B_lit=uf(49583e-9,40e-9)
a_lit=uf(66,1)*(2*np.pi)/360
out("\\sigma_{B_{E}, lit}",usigma(B_E,B_lit),"\\sigma")
out("\\sigma_{B_{E,V}, lit}",usigma(B_vert,B_lit*unp.sin(a_lit)),"\\sigma")
out("\\sigma_{B_{E,H}, lit}",usigma(B_hor,B_lit*unp.cos(a_lit)),"\\sigma")
out("\\sigma_{\\alpha, lit}",usigma(a*2*np.pi/360,a_lit),"\\sigma")
```

Wir sehen, dass alle Werte abgesehen vom Inklanationswinkel signifikant von den Literaturwerten abweichen. Dies deutet auf mindestens einen Systematischen Fehler in der Messung hin. Weiteres in der Diskussion. 

## Diskussion
Im ersten Versuchsteil haben wir das B-Feld der Helmholtzspulen mit der rotierenden Induktionsspule untersucht. Dabei haben wir eine signifikante ($4.15\sigma$) Abweichung zur aus dem Spulenstrom errechneten Feldstrke erhalten. Dies ist vermutlich auf mehrere Faktoren zurckzufhren. Zum einen gilt die Formel (4) nur im Zentrum der Spulen, die Induktionsspule besitzt aber durchaus nicht zu vernachlssigende Ausdehnung.  Zum anderen befindet sich in der Helmholtzspule eben auch noch die Induktionssplule mit allzu sparsamer Aufhngung. Andere, uere, Magnefelder knnten auch einen Einfluss genommen haben, jedoch ist eine Spekulation darber an dieser Stelle kaum zielfhrend ohne weitere Messungen. 

Anschlieend haben wir das Induktionsverhalten bei periodisch Wechselndem Feldstrom untersucht. Hier haben wir die Abhngigkeit der Induktionsspannung zum Winkel gegenber des Feldes aufgetragen und eine Abhngigkeit in Cosinusform gefunden, wie erwartet. Jedoch ist die Kurve um etwa 5 gegenber der erwarteten verschoben. Hier scheint die Skala oder die Spule um ebendiese 5 verschoben. In Diagramm 3 lsst sich dies betrachten. Wir haben die Induktivitt der Helmholtzspule untersucht, unser Messwert ist in einer hnlichen Grenordnung wie der theoretisch ermittlete Messwert. Jedoch wurde wahrscheinlich der Messfehler und der Fehler des theoretischen Wertes (Die Anleitung gibt keine Fehler zu den Spulendaten) unterschtzt, auch aus den oben bereits genannten Grnden, sodass sich eine signifikante Abweichung der Werte ergibt ($>10\sigma$). Der durch den Fit gefundene ohmsche Widerstand unterscheidet sich nicht signifikant von dem zuvor mit dem Multimeter gemessenen ($<2\sigma$). 

Zuletzt haben wir das Erdmagnetfeld vermessen (wollen). Wir finden jedoch nur den Inklanationswinkel in nicht signifikanter Abweichung von Literaturwerten. Die Messung des Kompenstionsstroms liefert Werte die sehr stark abweichen ($>14\sigma$). Die Messungen der Gesamt- und Horizontalfeldstrke unterscheiden sich zwar signifikant von den Literaturwerten, aber nicht ganz so stark ($<3.4\sigma$). Es ist unklar, woher diese Fehler genau stammen.

Die Fehlerbetrachtung dieses Versuches scheint sehr schwierig, da systematische Fehler vorzuliegen scheinen, welche jedoch nicht ohne greren Aufwand rechnerisch direkt geprft werden knnen.
