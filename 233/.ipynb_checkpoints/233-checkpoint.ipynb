## Grundlagen
### Beugungstheorie
Es gibt hier zwei Versuchsanordnungen, an welchen sich Beugung untersuchen lässt. 
#### Fesnelsche Beugung
Bei der Fresnelschen Beugung betrachtet man eine Lichtquelle und einen Schirm in endlichem Abstand vom Beugungsspalt, wie in Abb. 1 dargestellt. 


<center><b>Abbildung 1:  Fresnelsche Beugung (Quelle: Script)</b> </center>

Möchte man nun die Interferenz am beliebigen Punkt A auf dem Schirm betrachten, so wird schnell klar, dass dies kein komplett triviales Unterfangen ist, da die Lichtwege aus jedem Punkt der Beugungsebene auch  in einem anderen Winkel zu A gelangen. Wir wollen daher hier eine simplere Betrachtungsweise wählen:
#### Fraunhofersche Beugung
Hier erfahren wir die in Abb.2 gezeigte Vereinfachung, dass wir sowohl Lichtquelle als auch Beobachtungebene (optisch) ins Unendliche verschieben. Dieser Effekt lässt sich durch den Einsatz von Linsen  näherungsweise realisieren. 


<center><b>Abbildung 2:  Fraunhofersche Beugung (Quelle: Script)</b> </center>

In diesem Aufabu kann man nun nutzen, dass der Ausfallswinkel $\alpha$ zum Beobachtungspunkt konstant ist.

Wir wollen in diesem Versuch ausschließlich monochromatisches Licht betrachten, dieses habe die Wellenlänge $\lambda$. Nach dem Hygenschen Prinzip gilt, dass jeder Punkt im Beugungsspalt wieder eine Elementarwelle ausstrahlt. Aufgrund der monochromatie des Lichts gilt dann für die Welle im Spalt:

\begin{align}
E=E_0e^{i\omega t}
\end{align}
Für die Überlagerung im Beobachtungspunkt gilt dann 
\begin{equation}
E_{\infty}(\alpha)=\int_{-d / 2}^{+d / 2} E_{0} e^{i(\omega t-k l)} d y
\end{equation}
wobei $k=2 \pi / \lambda$ , $d$ die Spaltbreite , $y$ die koordinate im Spalt um den Ursprung  und l die Weglänge, sich ergebend aus
\begin{align}
l=R+y \sin \alpha
\end{align}
wobei sich $R$ wie in Abb.3 ergibt. 

<center><b>Abbildung 3:  Fraunhofersche Beugung am Spalt(Quelle: Script)</b> </center>

Werten wir das obige Integral nun mithilfe der Eulerschen Identität aus, so erhalten wir
\begin{equation}
\begin{aligned}
E_{\infty}(\alpha) &=E_{0} e^{i(\omega t-k R)} \frac{e^{-i k \sin \alpha d / 2}-e^{i k \sin \alpha d / 2}}{-i k \sin \alpha} \\
&=E_{0} e^{i(\omega t-k R)} \frac{\sin (\pi d \sin \alpha / \lambda)}{\pi \sin \alpha / \lambda}
\end{aligned}
\end{equation}
Sei
\begin{align}
x:= \frac{d}{\lambda} \pi \sin \alpha
\end{align}
Der obige Ausdruck vereinfacht sich mit $\sinc x = \frac{\sin x}{x}$ zu
\begin{align}
    E_{\infty}(\alpha) = e^{i(\omega t-k R)} \sinc (x) d 
\end{align}
Wollen wir die Intensität am Beobachtungspunkt ermitteln, so quadrieren wir dies. Vorfaktoren vernachlässige ich:
\begin{align}
I_{\infty}(x)\propto \sinc^2 x
\end{align}

### Herleitung der Fourierdarstellung der Fraunhoferschen Beugung

Wir wollen nun die Fraunhofersche Beugung an einem beliebig geformten zweidimensionalen Spalt errechenen. Wir wollen dafür allgemein für einen Punkt P die Interferenz bestimmen, seinen die Bezeichnungen wie in Abb. 4. 


<center><b>Abbildung 4:  Geometrie an beliebiger Öffnung (Quelle: Script)</b> </center>

Von jedem flächenelement $\d A$ geht eine Elementarwelle/Kugelwelle mit einer Anfänglichen Öffnungsfunktion $A(y,z)$:
\begin{equation}
E_{\mathrm{EW}}(r)=\frac{A(y, z)}{r} \cdot e^{i k r}
\end{equation}
wobei ich hier die Zeitabhängigkeit vernachlässige, da diese für die Intensität irrelevant ist. Wir wollen nun einige Abschätzungen machen um $r$ zu eleminieren. Zunächst lässt sich bei hinreichendem Abstand das $r$ im Nenner des Ausdrucks durch $R$ ersetzen. Für das $r$ im Exponenten machen wir folgende Abschätzung:
\begin{align}
r&=R \cdot \sqrt{1+\left(y^{2}+z^{2}\right) / R^{2}-2(Y y+Z z) / R^{2}} &\\
&\approx R \cdot \sqrt{1-2(Y y+Z z) / R^{2}} &\quad \text{da} \left(y^{2}+z^{2}\right) / R^{2}<<1\\
& \approx R\left(1-(Y y+Z z) / R^{2}\right) & \quad \text { Taylor in } 2(Y y+Z z) / R^{2}
\end{align}
Wir erhalten also folgende  Abschätzung:
\begin{equation}
E_{\mathrm{EW}}(R)=\frac{A(y, z)}{R} \cdot e^{i k(R-(Y y+Z z) / R)}
\end{equation}
Durch integration über die Beugungsebene erhält man nun die Welle am Punkt P zu
\begin{equation}
\begin{aligned}
E &=\int_{S} \frac{A(y, z)}{R} \cdot e^{i k(R-(Y y+Z z) / R)} \mathrm{d} \vec{A} \\
&=\frac{e^{i k R}}{R} \int_{S} A(y, z) e^{-i k(Y y+Z z) / R} \mathrm{~d} \vec{A}
\end{aligned}
\end{equation}
Da uns Normierung hier egal ist, werfen wir den Vorfaktor weg. Wir definieren
\begin{align}
k_y :&= k \frac{Y}{R} = k \sin \phi \\
k_z :&= k \frac{Z}{R} = k \sin \theta
\end{align}
Das Integral ist dann:
\begin{equation}
E\left(k_{y}, k_{z}\right)=\iint_{S} A(y, z) e^{-i\left(k_{y} y+k_{z} z\right)} d y d z
\end{equation}
Dies hat aber genau die Form einer zweidimensinalen Furiertransformierten! Die Feldverteilung der Beugungsstruktur unter Fraunhoferbeugung ist also die Fouriertransformierte der Öffnungsfunktion, anschaulich also gerade der Öffnung. Dies können wir nun nutzen um Beugungsstrukturen auszurechenen bzw. anzunähren.

### Beugungsbild des Einfachspalts

Wir interessieren uns hier nur für Beugung in y-Richtung. Für einen Spalt ist
\begin{equation}
A(y)= \begin{cases}1 & |y| \leq d / 2 \\ 0 & \text { sonst }\end{cases}
\end{equation}
mit Fouriertransformierten
\begin{equation}
F\left(k_{y}\right)=d \frac{\sin \left(k_{y} d / 2\right)}{\left(k_{y} d / 2\right)} \equiv \operatorname{sinc}\left(k_{y} d / 2\right) d
\end{equation}
mit Nullstellen
\begin{align}
k_y=2\pi n /d \quad n \in \mathbb{Z}
\end{align}
Rücktransformation ergibt das analytisch nicht lösbare Integral
\begin{equation}
\mathrm{f}(\mathrm{y})=\frac{1}{2 \pi} \int_{-\infty}^{\infty} F\left(k_{y}\right) e^{i k_{y} y} d k_{y}=\frac{d}{k} \int_{0}^{\infty}\left[\sin \left(k_{y} d / 2\right) /\left(k_{y} d / 2\right)\right] \cos \left(k_{y} y\right) d k_{y}
\end{equation}
Das Integral ist im endlichen Falle jedoch auswertbar, vergegenwärtigt man sich, dass steigende $k_y$ mit höherem Winkel korrespondieren, so kann man das "Abschneiden" des Integrals mit dem Abdecken in der Fourierebene identifizieren und experimentell beobachten. Siehe Abb. 5.


<center><b>Abbildung 5: Manipolation der Beugungsstruktur (Quelle: Script)</b> </center>

### Beugungsbild des Doppelspaltes 
Der Doppelspalt kann wie gewohnt als eine Überlagerung zweier Einfachspalte betrachtet werden. Es ergibt sich für die Fouriertransformierte: 
\begin{equation}
F = F(k_y,\text{rechts}) + F(k_y,\text{links}) = 2\cos(k_yg/2)d\frac{\sin(k_yd/2)}{(k_yd/2)}
\end{equation}
wobei $g$ der Spaltabstand ist. Wir sehen, dass die Funktion wie gewohnt die Fouriertrafo des Einzelspalts als Einhüllende besitzt. Substituieren wir nun $k_y = 2 \pi / \lambda \sin \alpha$ so erhalten wir:
\begin{equation}
I = 4\cos^2(\pi g /\lambda \sin\alpha)d^2\frac{\sin^2(\pi d/\lambda \sin\alpha)}{(\pi d/\lambda \sin\alpha)^2}
\end{equation}
und finden das "Abgeschnittene" Bild zu:
\begin{equation}
F_{mod}(y) \propto [f_{mod}(y)]^2 = [2d/\pi \int\limits_{0}^{k_y,n} \cos(k_y,g/2) \frac{\sin(k_yd/2)}{(k_yd/2)}\cos(k_yy)dk_y]^2
\end{equation} 
Hier lässt sich nun errechenen, ab wo die Doppelstruktur verschwindet, bei kontinuierlicher Verkeleinerung des Integrationsbereiches. Das Maximum erster Ordnung findet sich bei $\sin\alpha = \lambda/g$, von diesem aus kann man verkleinern.

#### Eichung
Wir müssen die Messung zunächst wie oben Beschrieben Eichen. Dafür tragen wir den Abstand der Analysierspalte zu den Pixeln auf. Die Steigung ist der Eichfaktor $a$:


```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
from IPython.display import Markdown, Latex, display
import pandas as pd
from scipy.optimize import curve_fit
import uncertainties
from uncertainties import ufloat as uf
from uncertainties.umath import *
from uncertainties import unumpy as unp
from uncertainties.unumpy import (nominal_values as noms,
                                  std_devs as stds)
from scipy.signal import *

table_counter=1

def round_up(n, decimals = 0):  
    multiplier = 10 ** decimals  
    return np.ceil(n * multiplier) / multiplier 

def out(name,c,unit="",zp=0,decimals=2):
    if isinstance(c, uncertainties.UFloat):
        display(Markdown("$"+name+" = " +'{:L}'.format(c)+" \   \mathrm{"+unit+"}$"))
    elif type(c) in [float,int]:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        print(c)
        
def df_out(df,title=""):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1

def usigma(g1,g2):
    return abs((g1.n-g2.n)/unp.sqrt(g1.s**2+g2.s**2))

def outwstd(name,c,Dc,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = (" +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+") "+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))

def out_sys_std(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        pass
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c[0]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+ "{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals) +" ) " + unit + " $"))
    else:
        display(Markdown( "$ "+name+" = ( " + "{:.{}f}".format(round(c[0]*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))
        
def out_no_error(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \cdot 10^{ "+str(zp)+" }  " + unit + " $"))

def df_out(df,title=""):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1
    
def sigma(g1,Dg1,g2,Dg2):
    return abs((g1-g2)/np.sqrt(Dg1**2+Dg2**2))

def quad_add(arr):
    quad_sum=0
    for i in arr:
        quad_sum+=i**2
    return (quad_sum**0.5)/len(arr)

def fit_line(x, m, n): 
    return m * x + n 

```

```python
data_eich = np.genfromtxt('eichung.csv', delimiter=",", unpack = True, skip_header=1) 
 
delx_eich = unp.uarray((data_eich[3]-data_eich[1]),np.sqrt(data_eich[2]**2 + data_eich[4]**2) )
 
dela = unp.uarray(2 * data_eich[5] ,2 * data_eich[6])
  
popt_eich, pcov_eich = curve_fit(fit_line, noms(delx_eich), noms(dela), sigma = np.sqrt(quad_add([stds(dela),stds(delx_eich)])))
                                                                                                                                         
eich = uf(popt_eich[0],np.sqrt(pcov_eich[0][0]))
     
x = np.linspace(250, 950, 2) 
 
plt.plot(x, fit_line(x, *popt_eich),label="Ausgleichsgerade") 
plt.grid(ls='dotted')
plt.errorbar(noms(delx_eich), noms(dela), yerr = stds(dela), xerr = stds(delx_eich), fmt = 'bx', capsize = 3)
 
plt.title('Eichkurve') 
plt.ylabel('Abstand[mm]') 
plt.xlabel('Abstand[pxl]') 
plt.legend() 
out("a",eich,"\\frac{mm}{px}")
```

#### Bestimmung der Spaltbreite
Wir wollen nun die Spaltbreite $d$ bestimmen. Dazu tragen wir zunächst den Abstand der Minima $s_{min}$ gegen die Ordnungszahl auf und bestimmen die Steigung $b$. Daraus ergibt sich dann 
\begin{align}
d&=\frac{n \lambda}{\sin \alpha} = \frac{ \lambda f}{b}\\
\Delta d&=d \sqrt{\left(\frac{\Delta f}{f}\right)^{2}+\left(\frac{\Delta b}{b}\right)^{2}}
\end{align}

Anschließend tragen wir in das gleiche Diagramm die Abstände der Maxima ein überprüfen, ob diese dann wie theoretisch gedacht genau zwischen den Ordnungen (also bei 1.5, 2.5, ...) liegen bzw. wie sehr davon abgewichen wird.


```python

data_1 = np.genfromtxt('v1.csv', delimiter=",", unpack = True) 
 
mask = np.isnan(data_1[5]) == False 
delx = 2*(data_1[1][0] - data_1[5][mask]) 
delx_err = 2 * np.sqrt(data_1[2][0]**2 + data_1[6][mask]**2) 
 
mask_2 = np.isnan(data_1[1]) == False 
delx_max = unp.uarray(2*(data_1[1][0] - data_1[1][mask_2]) , 2 * np.sqrt(data_1[2][0]**2 + data_1[2][mask_2]**2) )
 
popt_1, pcov_1 = curve_fit(fit_line, data_1[0][mask], delx, sigma = delx_err) 
 
n_max = (delx_max - popt_1[1]) / popt_1[0] 
 
x = np.linspace(0, 6, 2) 
 
plt.errorbar(data_1[0][mask], delx, yerr = delx_err, linewidth = 0, fmt = 'bx', capsize = 3, elinewidth = 2,  label="$s_{min}$")
plt.plot(x, fit_line(x, *popt_1), label = 'Ausgleichsgerade') 
plt.errorbar(noms(n_max) , noms(delx_max), yerr = stds(delx_max), linewidth = 0, fmt = 'rx', capsize = 3, elinewidth = 2,  label="$s_{max}$")

plt.title('$s_{min} in Abhängigkeit von der Beugungsordnung') 
plt.xlabel('$n$') 
plt.ylabel('Abstand [pxl]') 
plt.grid(ls='dotted')
plt.legend() 
plt.text(0, 470, 'm = ({0:.1f}±{1:.1f})px'.format(popt_1[0], np.sqrt(pcov_1[0][0]))) 
#plt.text(0, 400, 'n = ({0:.1f}±{1:.1f})pxl'.format(popt_1[1], np.sqrt(pcov_1[1][1])))

lam = 695 * 10**(-9)
f = uf(80,1)*10**(-3)
b = uf(popt_1[0], np.sqrt(pcov_1[0][0]))*eich * 10**(-3)
d=(lam*f)/b
out("b",b,"px")
out("d",d*10**(3),"mm")  
theo_n_max=[1.5,2.5,3.5,4.5,5.5]
df_espalt=pd.DataFrame(data={"Berechnete Ordnung $n$":n_max[1:],"Abweichung":abs(noms(n_max[1:])-theo_n_max)})
df_out(df_espalt,"Berechnete Ordnungen der Maxima")
```

#### Intensitätsverteilung der Maxima
Wir wollen nun die Intensitäten der Maxima vergleichen mit dem theoretischen Werten. Wir beziehen uns dafür immer relativ zum ersten Maximum. Theoretisch gilt:
\begin{align}
\left(\frac{I}{I_{0}}\right)_{\text {theo }}&=\sinc^{2}(x)\\
%\Delta\left(\frac{I}{I_{0}}\right)_{\text {theo }}&=\frac{x \sin (2 x)+\cos (2 x)-1}{x^{3}} \Delta x
\end{align}
Ist U die Untergundintensität, so ist diese im Experiment abzuziehen:
\begin{align}
\left(\frac{I}{I_{0}}\right)_{\exp }&=\frac{I-U}{I_{0}-U}\\
\Delta\left(\frac{I}{I_{0}}\right)_{\exp }&=\sqrt{\left(\frac{\Delta I}{I_{0}-U}\right)^{2}+\left(\frac{I_{0}-I}{\left(I_{0}-U\right)^{2}} \Delta U\right)^{2}+\left(\frac{I-u}{\left(I_{0}-U\right)^{2}} \Delta I_{0}\right)^{2}}
\end{align}


```python
data_ug = np.genfromtxt('v1_ug.csv', delimiter=",", unpack = True) 
max_2 = 116 
max_2_err = 3 
 
I_ug_1 = uf(np.mean(data_ug[2][:1]), np.sqrt(np.sum(data_ug[2][:1]**2)) / 2 )
I_ug_2 = uf(np.mean(data_ug[2][1:]) ,np.sqrt(np.sum(data_ug[2][1:]**2)) / 2 )
 
I_rel = np.empty(len(data_1[3][mask_2]),dtype=object)
I_rel[:2] = (data_1[3][mask_2][:2] - I_ug_1) / (data_1[3][0] - I_ug_1) 
I_rel[2:] = (data_1[3][mask_2][2:] - I_ug_2) * (max_2 - I_ug_1) / (data_1[3][mask_2][2] - I_ug_2) / (data_1[3][0]-I_ug_1)
 
I_theo = [1, 0.047190, 0.016480, 0.008340, 0.005029, 0.003361] 

df_int_es=pd.DataFrame(data={"$I_{theo}$":I_theo,"$I_{exp}$":I_rel,"$\sigma$":sigma(noms(I_rel),stds(I_rel),I_theo,0.1*10**(-99))})

df_out(df_int_es,"Theoretische und experimentelle Intensitäten im Vergleich")
```

### Beugungsstruktur des Doppelspaltes
Wir wollen auch hier wieder Theorie mit Experiment vergleichen. Für die Berechnung der theoretischen Werte nutzen wir wie oben hergelietet
\begin{align}
I(x)=\sinc^2 (x) \cos^2 (\frac{g}{d}\pi x)
\end{align} 
Ein Blick in die Messdaten zeigt, dass dies tatsächlich dem folgenden Diagram sehr ähnlich aussieht. Daran lässt sich auch gut erkennen, welche minima durch welchen Therm zustande kommen. 

```python
#Pixel-positionen der Spalte aus Aufgabe 4
%matplotlib inline
breite1= uf(143,8)
breite2= uf(138,8)
min_abstand= uf(218,8)
d_d=np.mean([breite1,breite2])

g= 0.5*breite1+0.5*breite2+min_abstand
v=g/d_d

def beugung_doppelspalt(x):
    return np.sinc(x)**2*np.cos(np.pi*v.n*x)**2

x=np.linspace(-3,3,200001)
plt.plot(x,beugung_doppelspalt(x),color='blue',label='Doppelspalt')
plt.plot(x,np.sinc(x)**2,color='darkblue',label='Einzelspalt',linestyle='--')
plt.title('Beugung am Spalt und Doppelspalt')
plt.xlabel('$\pi \cdot x$')
plt.ylabel('relative Intensität  [b.E.]')
plt.grid(ls='dotted')
plt.ylim(0,1.1)
plt.xlim(-2.5,2.5)
plt.text(-2,0.92,r'$v = 2.55$',size=13)
_=plt.legend(frameon=True)
```
