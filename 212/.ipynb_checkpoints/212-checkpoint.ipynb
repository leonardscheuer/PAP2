## Motivation/Versuchsziel

Wir wollen die **Viskosität** $\eta$ einer Flüssigkeit bestimmen. Diese beschreibt die umgangssprachliche "Zähigkeit" oder "Dickflüssigkeit". Spezifisch soll hier die Viskosität von  Polyethylenglykol (kurz PEG) mit verschiedenen Methoden bestimmt werden. Umgekehrt soll das Gesetzt von Stokes in laminaren und turbolenten Situationen verglichen werden. 

## Grundlagen

Ich möchte zunächst einen Körper betrachten, der sich mit konstanter Geschwindigkeit durch ein viskoses MEdium bewegt. Es ist eine Kraft erforderlich, um die Geschwindigkeit konstant zu halten, denn im Medium wirkt auf einen bewegten Körper eine Reibugskraft $F_R$ entgegen der Bewegungsrichtung, die durch zwischenmolekulare Kräfte verursacht wird.
Die Bewegung des Körpers bewirkt, dass sich ein umgebender Teil des Mediums mit ihm bewegt.
Es gilt das **Newtonsches Reibungsgesetz**:

\begin{align}
	F_r =& \eta A \frac{v}{z}
\end{align}
Bewegen sich die Schichten nicht mit linearem Strömungsgradienten, so ist allgeminer:
\begin{align}
	F_r =& \eta A \frac{\text{d}v}{\text{d}z} \label{eq:NR}
\end{align}


Dabei beschreibt $\eta$ die Viskosität, $A$ die Fläche des als Scheibe genährten Körpers, $v$ die Geschwindigkeit und
$z$ den Abstand zu einer ruhenden Platte, an welcher die Flüssigkeit ruht. Es wird angenommen, dass sich die Flüssigkeit linear schneller in Schichten fortbewegt. Eine solche Strömung ist unturbulent und wird daher auch als **laminar** bezeichnet. Die Flüssigkeitsschichten vermischen sich dabei nicht. 
Für eine Kugel mit einem Radius $r$, Geschwindigkeit $v$ und $z\rightarrow \infty$ (Medium unbegrenzt) gilt im laminaren Fall das **Stoke'sche Reibungsgesetz**:
\begin{align}
	F_r=& 6\pi \eta rv \label{eq:SR}
\end{align}


Verläuft eine Strömung in Umgebung des Körpers nicht mehr in getrennten Schichten, z.B. wenn wirbel auftreten, wie es häufig bei größeren Geschwindgkeiten oder "unschnittigen" Körperformen auftritt mit $F_R$ sehr viel größer, so nennen wir diese **turbulent**.
Um laminare und turbulente Strömung miteinander Vergleichen zu können, gibt es die **Reynoldszahl**:
\begin{align}
	\text{Re}=&\frac{2E_{\text{kin}}}{W_r}= \frac{\rho v L}{\eta} \label{eq:Re}
\end{align}
wobei $E_{\text{kin}}$ die kinetische Energie und $W_r$ die Reibungsverluste sind. Mit der mittleren Strömungsgeschwindigkeit $v$ und einer charakteristischen Länge $L$ erhält man die obenstehende Gleichheit. Im Falle eines Rohresist $L$ der Rohrdurchmesser. Ist die Rynoldszahl klein, so handelt es sich um eine Laminare Strömung, ist sie groß um eine turbulente. Die kritische Reynoldszahl ist jener Wert, wo die laminare Strömung instabil und schließlich turbulent wird.

### Hagen-Poiseuille: Laminare Rohrströmung
Eine Andere Methode zur Viskositätsbestimmung ist die Messung des Volumenstroms durch ein Rohr mit gegebenem Durchmesser, in welchem eine Laminare Strömung stattfindet. 
Wir betrachten ein Rohr mit der Länge $L$ und dem Radius $R$. Durch die Druckdifferenz $\Delta p$ zwischen Anfang und Ende des Rohres (bzw. des Betetrachteten Rohrelements) entsteht eine Strömung. Dadurch wirkt eine Kraft:
\begin{align}
	F_p =& \pi r^2 \Delta p
\end{align}
Durch Reibung mit dem Rohr entsteht außerdem eine Reibungskraft:
\begin{align}
	F_r =&  \eta \cdot 2 \pi r L \left(-\frac{\text{d}v}{\text{d}r}\right)
\end{align}
Durch Gleichsetzten und Integrieren (unter Beachtung der Randbedingungen) ergibt:
\begin{align}
	v(r) =& \frac{R^2-r^2}{4L\eta}\Delta p
\end{align}
Daraus ergibt sich der Volumenstrom
\begin{align}
	\frac{\text{d}V}{\text{d}t} = \frac{\pi R^4}{8L\eta}\Delta p
\end{align}
und schließlich das **Gesetz von Hagen-Poiseuille**:
\begin{align}
 \eta =&  \frac{\pi R^4}{8L\frac{\text{d}V}{\text{d}t} }\Delta p \label{eq:HP}
 \end{align}
 Die Kritische Reynoldszahl ist hier etwa 2300. Aus dem Pascalschen Gesetzt erhalten wir zudem:
 \begin{align}
	\Delta p =& g\rho \Delta h \label{eq:p}
\end{align}
wobei $\Delta h$ die Höhe der Flüssigkeit über dem Auslass ist.

## Auswertung
$\newcommand{\lr}[1]{\left(#1\right)}$
Wir wollen nun die Viskosität aus den Messwerten des ersten Versuchsteils ermitteln.
Zunächst stellen wir (5) um zu:
\begin{align}
	m =& \frac{2g}{9\eta}
   \end{align}
Wobei $m$ die Steigung der Geraden ist, wenn $l:=\frac{v}{\rho_k-\rho_f}$ auf $r^2$ aufgetragen wird. Wir finden
\begin{align}
\Delta \eta =& \frac{2g}{9m^2}\Delta m.
\end{align}
Um die Fehler der Datenpunkte von $l$ zu berechenen findet man:
\begin{align}
\Delta l = \frac{s}{\lr{\frac 1n\sum\limits_{i=1}^n t_i}\cdot \lr{\rho_{k}-\rho_{f}}} \sqrt{\lr{\frac{\Delta s}{s}}^2 + \frac{\sum\limits_{i=1}^n \Delta t_i^2}{\lr{\sum\limits_{i=1}^n t_i}^2} + \frac{ \Delta \rho_{k}^2+\Delta \rho_{f}^2}{\lr{\rho_{k}-\rho_{f}}^2}}
\end{align}
Wobei $s$ die Fallstrecke ist und die $t_i$ die Fallzeiten. 
Die in den Formeln auftretenen Fehler sind im Messprotokoll vermerkt. 
Wegen der endlichen Ausdehnung des Rohres ist die Annahme der Stoke'schen Reibung nicht komplett korrekt. Ausgeleichend können wir die Ladenburgsche Korrektur verwenden, indem wir $F_R$ und daraus folgend die Fallgeschwindigkeiten mit

\begin{equation}
\lambda=1+2,1 \frac{r}{R}
\end{equation}

multiplizieren. Es ist

\begin{equation}
\Delta \lambda=2,1 \sqrt{\left(\frac{\Delta r}{R}\right)^{2}+\left(\frac{r}{R^{2}} \Delta R\right)^{2}}.
\end{equation}


Wir tragen nun $l$ auf, bestimmen daraus $m$ und $\eta$, letztere für die Korrigierten Werte:

```python
#Funktionen zur späteren Verwendung. 
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
from IPython.display import Markdown, display, Latex
import pandas as pd
from scipy.optimize import curve_fit
table_counter=1

def comma_to_float(valstr):
  return float(valstr.decode("utf-8").replace(',','.'))

def round_up(n, decimals = 0):  
    multiplier = 10 ** decimals  
    return np.ceil(n * multiplier) / multiplier 

def outwstd(name,c,Dc,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = (" +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+") "+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))

def out_sys_std(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        pass
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c[0]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+ "{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals) +" ) " + unit + " $"))
    else:
        display(Markdown( "$ "+name+" = ( " + "{:.{}f}".format(round(c[0]*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))
        
def out_no_error(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))

def df_out(df,title):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1
    
def sigma(g1,Dg1,g2,Dg2):
    return abs((g1-g2)/np.sqrt(Dg1**2+Dg2**2))

def quad_add(arr):
    quad_sum=0
    for i in arr:
        quad_sum+=i**2
    return (quad_sum**0.5)/len(arr)

def linear(x,a,b):
    return a*x+b

def fit_line(x,m): 
    return m*x
```

```python
kf = np.genfromtxt('Kugelfall.csv', converters = {0:comma_to_float, 1:comma_to_float, 2:comma_to_float, 3:comma_to_float, 4:comma_to_float, 5:comma_to_float, 6:comma_to_float, 7:comma_to_float}, delimiter = ';', unpack=True)
df = pd.DataFrame(data=kf, columns=["r[mm]", "Fallstrecke[mm]","Fallzeit 1[s]","Fallzeit 2[s]","Fallzeit 3[s]","Fallzeit 4[s]","Fallzeit 5[s]"])
df["r[mm]"]=df["r[mm]"]/2

mean_v = np.zeros(len(kf)) 
mean_err = np.zeros(len(kf)) 
r = np.zeros(len(kf)) 
dens = np.zeros(len(kf)) 
dens[:-2] = 1378 
dens[-2] = 1358 
dens[-1] = 1363 
dens_l = 1148.4 # density of liquid, in SI units 
dens_l_err = 0.3 
dens = dens - dens_l 
dens_err = np.sqrt(3**2 + dens_l_err**2) 

R = 0.075/2 #durchmesser der Röhre
R_err = 0.001/2 

#%matplotlib notebook
%matplotlib inline

for i in range(0, len(kf)):
    mean_v[i] = kf[i][1]*10**(-3)/np.mean(kf[i][2:])
    mean_err[i] = len(kf[i][2:])*0.2**2/(np.sum(kf[i][2:])**2)
    
r=df["r[mm]"]*(10**(-3))
    
vdevrho = mean_v / dens
vdevrho_err = vdevrho * np.sqrt((mean_err/mean_v)**2 + dens_err/(dens**2))
vdevrho_corr = mean_v*(1+2.1 * r/R) / dens
vdevrho_corr_err = vdevrho_corr * np.sqrt((vdevrho_err/vdevrho)**2+(2.1 * r / R/(1 + 2.1 * r / R))**2 * (0.01**2 + R_err**2/R**2))

plt.errorbar(r**2, vdevrho, xerr = (0.01 * r)**2, yerr = vdevrho_err, fmt = 'bx', label = 'Werte ohne Korrektur', capsize = 3)
plt.errorbar(r**2, vdevrho_corr, xerr = (0.01 * r)**2, yerr = vdevrho_corr_err, fmt = 'rx', label = 'Werte mit Korrektur', capsize = 3)
plt.xlabel('$r^2 [m^2]$')
plt.ylabel("$ \\frac{v}{\\rho_k-\\rho_f} [\\frac{m^4}{kg \cdot s}]$ ")

popt_corr, pcov_corr = curve_fit(fit_line, (r**2) , vdevrho_corr, sigma = vdevrho_corr_err)
popt, pcov = curve_fit(fit_line, (r**2) , vdevrho, sigma = vdevrho_err)


x = np.linspace(0,2.25*10**(-5), 2)
plt.plot(x, fit_line(x, popt_corr[0]), color="red" ,label = 'lin. Fit der korrigierten Werte')
plt.plot(x, fit_line(x, popt[0]), color="blue" ,label = 'lin. Fit der nicht korrigierten Werte')
plt.title('$l$ in Abhängigkeit von $r^2$')
_=plt.legend()


g = 9.80984
eta = 2*g/(9*popt_corr[0])
err_eta = 2*g/(9*popt_corr[0]**2)*pcov_corr[0][0]**0.5
outwstd("\eta",eta,err_eta,zp=0,decimals=3,unit="\\frac{kg}{sm}")

df["$\\bar{v} [mm/s]$"]= df["Fallstrecke[mm]"]/df.drop(columns=["r[mm]", "Fallstrecke[mm]"]).mean(axis=1)

stds=pd.Series([],dtype='float64')
for i in range(0,len(df["Fallstrecke[mm]"])):
    stds[i]=(df["Fallstrecke[mm]"][i]/df.drop(columns=["r[mm]", "Fallstrecke[mm]","$\\bar{v} [mm/s]$"]).loc[i]).std(ddof=0,axis=0)

df["$\\Delta \\bar{v} [mm/s]$"]= stds
    
#df["$\\Delta \\bar{v} [mm/s]$"]=  (df["Fallstrecke[mm]"][i]/df.drop(columns=["r[mm]", "Fallstrecke[mm]"])).std()
#df_out((df["Fallstrecke[mm]"]/df.drop(columns=["r[mm]", "Fallstrecke[mm]","$\\bar{v} [mm/s]$"])),"")
#df_out(df,"")
```

Wir wollen die Geschwindigkeiten nun Vergleichen mit den Theoretisch aus (5) bzw. der gefunden Geraden zu erhaltenden. Es ergibt sich aus (5) der Fehler:
$\renewcommand{\t}[1]{\text{#1}}$
\begin{align}
	\Delta v_{\t{lam}} =& v_{\t{lam}} \sqrt{\frac{\Delta \rho_{k}^2+\rho_{l}^2}{(\rho_{k}-\rho_{l})^2} + 
	\frac{2,1\frac{r}{R}(\lr{\frac{\Delta r}{r}}^2 + \lr{\frac{\Delta R}{R}}^2)}{(1+2,1\frac rR)^2}
	+\lr{2\frac{\Delta r}{r}}^2 + \lr{\frac{\Delta \eta}{\eta}^2}
	}
\end{align}
Daraus wollen wir pro Messpunkt die Reynoldszahl nach (4) berechenen, der Fehler Ergibt sich zu:
\begin{align}
	\Delta\text{Re} =& \t{Re}\cdot \sqrt{\lr{\frac{\Delta \rho}{\rho}}^2 + \lr{\frac{\Delta v}{v}}^2 + \lr{\frac{\Delta \eta}{\eta}}^2+\lr{\frac{\Delta r}{r}}^2}
\end{align}
Wollen wir nun $\t{Re}_{krit}$ bestimmen, so tragn wir $\bar{v}/v_{lam}$ zu $\log ( \t{Re})$ auf und bestimmen dann den "Knickpunkt" im Plot. Bei uns befindet sich dieser zwischen dem 3. und 4. Messpunkt. Zum Schluss gebe ich noch eine Übersichtstabelle aus.


```python
df["$v_{lam} [mm/s]$"]= popt_corr[0] * (r**2) * dens *10**3
df["$\Delta v_{lam} [mm/s]$"] = df["$v_{lam} [mm/s]$"] * np.sqrt((pcov_corr[0][0]**0.5/popt_corr[0])**2 + (2*0.01)**2 + dens_err/(dens**2))
df["$\sigma$"]=sigma(df["$v_{lam} [mm/s]$"],df["$\\Delta v_{lam} [mm/s]$"],df["$\\bar{v} [mm/s]$"],df["$\\Delta \\bar{v} [mm/s]$"])
df["Re"]=1148.4 * df["$\\bar{v} [mm/s]$"]*10**(-3) * 2 * r / eta
df["$\Delta$Re"]=df["Re"] * np.sqrt((0.3/1148.4)**2 + (df["$\\Delta \\bar{v} [mm/s]$"]/df["$\\bar{v} [mm/s]$"])**2 + 0.01**2 + (err_eta/eta)**2)
df_out(df.drop(columns=["Fallstrecke[mm]","Fallzeit 1[s]","Fallzeit 2[s]","Fallzeit 3[s]","Fallzeit 4[s]","Fallzeit 5[s]"]).round(3),"Übersicht über Geschwindigkeitswerte und Reynoldszahlen")

plt.errorbar(df["Re"], df["$\\bar{v} [mm/s]$"]/df["$v_{lam} [mm/s]$"], xerr = df["$\Delta$Re"], yerr = df["$\\bar{v} [mm/s]$"]/df["$v_{lam} [mm/s]$"] * np.sqrt((df["$\Delta v_{lam} [mm/s]$"]/df["$v_{lam} [mm/s]$"])**2+(df["$\\Delta \\bar{v} [mm/s]$"]/df["$\\bar{v} [mm/s]$"])**2), fmt = 'bx', capsize = 3)
plt.xscale('log')
plt.xlabel('Re')
plt.ylabel('$\\bar{v} /v_{lam}$')
_=plt.title('$\\bar{v} /v_{lam}$ zu Re')

Re_crit = (df["Re"][3] + df["Re"][4]) / 2
Re_crit_err = max(Re_crit - df["Re"][3], df["Re"][4] - Re_crit)
outwstd("\\text{Re}_{krit}",Re_crit, Re_crit_err,0,2)
```

## Hagen-Poiseuille

Wir wollen nun die Viskosität über Hagen-Poiseuille (wie oben beschrieben) bestimmen. Dafür wollen wir zunächst den Volumenfluss bestimmen. Dies geschieht im nachfolgenden Fit. Die inverse Steigung der Geraden ist dann der Volumenfluss. Der Fehler ergibt sich direkt daraus. Außerdem brauchen wir noch die (mittlere) Druckdifferenz $\bar{\Delta p}$. Mit Anfangshöhe $h_a$ und Gesamtvolumen $V$ ergibt sich:
\begin{align}
	\overline{\Delta p} =& \frac{\rho g (h_a+\:h_a-\frac{V}{\pi R^2})}2\\
	=& \rho g (h_a-\frac 12\frac{V}{\pi R^2})
\end{align}
Mit dem Kapillarenradius $R_K$ und der Kapillarlänge $L_K$ ergibt sich eingesetzt:
\begin{align}
	\eta  =& \frac{\pi \cdot R_K^4}{8 L \cdot \frac{\text{d}V}{\text{d}t} }\cdot \rho g (h_a-\frac 12\frac{V}{\pi R^2})\\
  \Delta \eta =& \eta \sqrt{\lr{\frac{\Delta L}{L}}^2+\lr{4\frac{\Delta R_K}{R_K}}^2+\lr{\frac{\Delta\lr{\frac{\t{d}V}{\t{d}t}}}{\frac{\t{d}V}{\t{d}t}}}^2+\lr{\frac{\Delta \rho}{\rho}}^2+\frac{\Delta h_a^2 + \lr{\frac{V}{2\pi R^2}}^2\lr{\lr{\frac{\Delta V}{V}}^2+\lr{2\frac{\Delta R}{R}}^2}}{ (h_a-\frac 12\frac{V}{\pi R^2})^2}}
\end{align}
Wir können nun die Viskosität errechenen. Schließlich vergleichen wir mit dem ersten Wert.




```python
hp = np.genfromtxt('Hagen_Poiseuille.csv', converters = {0:comma_to_float, 1:comma_to_float, 2:comma_to_float}, delimiter = ';', unpack = True)
hp[0] *= 10**-6 #SI units
plt.errorbar(hp[0], hp[1], yerr = hp[2], fmt = 'bx', capsize = 3)
popt2, pcov2 = curve_fit(linear, hp[0] , hp[1], sigma = hp[2])
x2 = np.linspace(0,3 *10**(-5),2)

plt.plot(x2, linear(x2 , popt2[0],popt2[1]), label = 'Ausgleichsgerade')
plt.ylabel('t [s]')
plt.xlabel('Volumen $[m^3]$')
#plt.legend()
_=plt.title('Durchgeflossenes Volumen $V$ in Abhängigkeit zur Zeit $t$')
dVdt = 1/popt2[0]
dVdt_err = pcov2[0][0]**0.5/(popt2[0]**2) #Error calculated with Gauß
h_a = 0.525
h_a_err = 0.001
V_end = 30 * 10**(-6)
R_K = 0.5*1.5*10**(-3)
R_K_err = 0.5*0.01*10**(-3)
L_K = 0.1
L_K_err = 5 * 10**(-4)
eta2 = np.pi * R_K**4 /(8 * dVdt * L_K) * dens_l * g * (h_a - 0.5 * V_end / (np.pi * R**2))
eta2_err = eta2 * np.sqrt((L_K_err / L_K)**2 + (4 * R_K_err / R_K)**2 + (dVdt_err / dVdt)**2 + (dens_l_err / dens_l)**2 + (h_a_err**2 + (V_end * R_err / (np.pi * R**3))**2)/((h_a - 0.5 * V_end / (np.pi * R**2))**2))
outwstd("\\eta",eta2, eta2_err,zp=0,decimals=3,unit="\\frac{kg}{sm}")
out_no_error("\sigma",sigma(eta,err_eta,eta2,eta2_err),zp=0,decimals=2,unit="")
```

Wir finden also eine  gute Übereinstimmung der Werte. Wir wollen nun wieder die Reynoldszahl bestimmen. Da Volumenstrom pro Fläche der Geschwindigkeit entspricht, ergibt sich aus (4): 
\begin{align}
	\t{Re} =& \frac{2 \rho \cdot \frac{\t{d}V}{\t{d}t}}{\eta \cdot \pi R}\\
	 \Delta \t{Re} =& \t{Re}\sqrt{\lr{\frac{\Delta \rho}{\rho}}^2+\lr{\frac{\Delta \frac{\t{d}V}{\t{d}t}}{\frac{\t{d}V}{\t{d}t}}}^2 + \lr{\frac{\Delta \eta}{\eta}}^2 + \lr{\frac{\Delta R}{R}}^2}
\end{align}
Wir erhalten:

```python
Re2 = 2*dens_l*dVdt / (eta2 * np.pi * R)
Re2_err = Re2 * np.sqrt((dens_l_err/dens_l)**2 + (dVdt_err/dVdt)**2 + (eta2/eta2_err)**2 + (R_err/R)**2)
outwstd("\\text{Re}",Re2, Re2_err,0,3)
```

Wir sehen, dass diese Zahl sehr klein ist (<<2300), also handelt es sich um eine laminare Strömung, wie für die Methode notwendig.

## Diskussion

Die Viskosität des PEGs im Praktikumsaufbau ist hier auf zwei verschiedene Arten bestimmt worden. Beide Arten ergeben nicht signifikant abweichende Ergebnsisse ($<1.5 \sigma$). In ersterer Methode haben wir auf alle Datenpunkte die ladenburgsche Korrektur angewandt um den Fehler duch die endliche Ausdehnung der Flüssigkeit zu kompensieren. Zweitere Methode hat einen etwas größeren Fehler. Wir vergleichen daher die Beiträge der verschiednen Summanden in (22), gerundet auf fünf Nachkommastellen:

```python
out_no_error("(\\frac{\Delta L}{L})^2",(L_K_err / L_K)**2,0,5,"")
out_no_error("\\lr{4\\frac{\Delta R_K}{R_K}}^2",(4 * R_K_err / R_K)**2,0,5,"")
out_no_error("\lr{\\frac{\Delta\lr{\\frac{\t{d}V}{\t{d}t}}}{\\frac{\t{d}V}{\t{d}t}}}^2",(dVdt_err/dVdt)**2,0,5,"")
out_no_error("\lr{\\frac{\Delta \\rho}{\\rho}}^2",(dens_l_err / dens_l)**2,0,5,"")
out_no_error("\\frac{\Delta h_a^2 + \lr{\\frac{V}{2\pi R^2}}^2\lr{\lr{\\frac{\Delta V}{V}}^2+\lr{2\\frac{\Delta R}{R}}^2}}{ (h_a-\\frac 12\\frac{V}{\pi R^2})^2}",(h_a_err**2 + (V_end * R_err / (np.pi * R**3))**2)/((h_a - 0.5 * V_end / (np.pi * R**2))**2),0,5,"")
```

Wir sehen, dass zweiterer Therm, der relative Fehler des Radius der Kapillare, mit Abstand den größten Fehler produziert. Eine genauere Angabe des Kapillarradius würde hier also den Fehler verkleinern. Die Ursachen der Fehler in ersterem Versuchsteil ist etwas schwerer zu sehen, da der Fehler der Viskosität proportional zum Fehler der Geraden im Fit ist, der Numerisch ermittelt ist. Allerdings wurden im Fit alle Datenpunkte verwendet, obwohl das Modell nur Laminare Strömungen berücksichtigt. Vergleicht man hinterher die Reynoldszahlen der Messwerte, so würde man besser die letzten drei Messwerte vernachlässigen, da diese über der Kritischen Reynoldszahl liegen. Tun wir dies (siehe unten), so nären sich die Ergebnisse beider Verfahren an. Oben in der Auswertung wurde dies noch nicht getan, da dort an dieser Stelle, vor der bestimmung der Reynoldszahlen, nicht eindeutig ersichtlich war, dass diese Punkte kein lineares Verhalten aufweisen. Aufgrund der Erfahrungen bei der Messdurchführung scheint der Fehler der Zeit (vor allem bedingt durch die Reaktionszeiten) recht groß, eventuell könnte man hier durch eine elektronische Messung (z.B eine Lichtschranke) bessere Fehler erhalten. 
Leider können alle erhaltenen Messwerte nicht mit Literaturwerten Verglichen werden, da PEG je nach Zusammensetztung verschiedene Viskosität besitzt und die Zusammensetztung im Praktikumsaufbau nicht näher bekannt ist.
