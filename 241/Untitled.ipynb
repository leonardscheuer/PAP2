## Grundlagen 
### Zeitentwicklung einer RC-Gliedes

Ein RC-Glied sei eine Schaltung wie in Abb. 1. ber den Widerstand wird der Kondensator geladen, bis schleilich die Angelegte Gleispannung erreicht wird. Nutzt man $I=\dot{Q}=C\dot{U_C}$ und die Kirchhoffsche Maschenregel, so erhlt man als Lsungen der entstehenden Differentialgleichung:
\begin{align}
U_R&=U_Ee^{-t/\tau}\\
I&=I_0e^{-t/\tau}
\end{align}
wobei $\tau=RC$. Experimentell kann $\tau$ bestimmt werden, indem die Halbwertszeit gemessen wird, es ist dann
\begin{align}
\tau=\frac{T_{12}}{\ln 2}
\end{align}

### Impedanz

Betrachten wir nun einen Wechselstrom, welcher zur Vereinfachung komplexwertig sei (in bekanntem Modell). Dann ist die Impedanz die Verallgemeinerung des Widerstandes als 
\begin{align}
Z=\frac{U}{I}.
\end{align} Fr die hier genutzten Bauteile ergibt sich:
\begin{align}
Z_R&=R\\
Z_C&=-i/(\omega C)\\
Z_L&=i\omega L
\end{align}
Rein imaginre Impedanzen nennt man auch Blindwiderstnde, sie verbrauchen keine Leistung und bewirken nur einen Phasenshift zwischen Strom und Spannung.
