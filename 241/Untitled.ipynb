## Grundlagen 
### Zeitentwicklung einer RC-Gliedes

Ein RC-Glied sei eine Schaltung wie in Abb. 1. Über den Widerstand wird der Kondensator geladen, bis schleißlich die Angelegte Gleispannung erreicht wird. Nutzt man $I=\dot{Q}=C\dot{U_C}$ und die Kirchhoffsche Maschenregel, so erhält man als Lösungen der entstehenden Differentialgleichung:
\begin{align}
U_R&=U_Ee^{-t/\tau}\\
I&=I_0e^{-t/\tau}
\end{align}
wobei $\tau=RC$. Experimentell kann $\tau$ bestimmt werden, indem die Halbwertszeit gemessen wird, es ist dann
\begin{align}
\tau=\frac{T_{12}}{\ln 2}
\end{align}


Abb. 1: links: RC-Glied, rechts: Spannung und Strom am Kondensator

### Impedanz

Betrachten wir nun einen Wechselstrom, welcher zur Vereinfachung komplexwertig sei (in bekanntem Modell). Dann ist die Impedanz die Verallgemeinerung des Widerstandes als 
\begin{align}
Z=\frac{U}{I}.
\end{align} Für die hier genutzten Bauteile ergibt sich:
\begin{align}
Z_R&=R\\
Z_C&=-i/(\omega C)\\
Z_L&=i\omega L
\end{align}
Rein imaginäre Impedanzen nennt man auch Blindwiderstände, sie verbrauchen keine Leistung und bewirken nur einen Phasenshift zwischen Strom und Spannung.

### Frequenzverhalten von RC-Gliedern

Legt man an ein RC-Glied nun also eine Wechselspannung an, so addieren sich die Impedanzen wie bei einer normalen Reihenschaltung. Über dem Kondensator erhält man so die Spannung
\begin{align}
U_C=\frac{Z_C}{R+Z_C}U_E
\end{align} 
Wodraus sich Betrag und Phase bestimmen lassen:
\begin{equation}
\begin{array}{c}
\left|U_{C}\right|=\frac{\left|U_{E}\right|}{\sqrt{1+(\omega R C)^{2}}} \\
\tan \varphi=-\omega R C
\end{array}
\end{equation}
Wird stattdessen über dem Widerstand abgegriffen, so haben wir einen Hochpass gebaut. Analog finden wir Betrag und Phase (siehe Abb. 2). Die enstrechenden Frequenzgänge sind in Abb. 2. sichtbar.

### RC-Glied als Differenzier- und Integrierglied


Abb. 3: Tiefpass als Integrator, Hochpass als Differenziator für $\tau>>T$. 

RC-Glieder können wenn passend gewählt ($\tau>>T$) (s. auch Abb.3) auch als Differenzier- oder Intergralglied eingesetzt werden. Für den Hochpass lässt sich dies quantitativ beschreiben als
\begin{equation}
\frac{d U_{A}}{d t}=\frac{1}{R C}\left(U_{E}-U_{A}\right)
\end{equation}
Für den Tiefpass:
\begin{equation}
U_A=\frac{1}{R C}\int U_{E} dt
\end{equation}

### Schwingkreis 
Ein Serienschwingkreis ist eine RCL-Reihenschaltung. Durch die Spule entsteht ein gegenläufiger Strom, wenn der Kondensator sich entlädt. Der Widerstand sorgt für eine Dämpfung, sodass, wie bekannt, das Problem als harmonischer Oszillator beschrieben werden kann. Als allegmeine Lösung findet sich:
\begin{equation}
I(t)=I_{o} e^{-\frac{R}{2 L} t}\left(c_{1} e^{\sqrt{\frac{R^{2}}{4 L^{2}}-\frac{1}{L C} t}}+c_{2} e^{-\sqrt{\frac{R^{2}}{4 L^{2}}-\frac{1}{L C} t}}\right)
\end{equation}
Für die Resonanzfrequenz ergibt sich: 
\begin{equation}
\omega_R=\sqrt{\frac{1}{LC}}
\end{equation}
Im Resonanzfall verhält sich der Schwinkreis so, als bestünde er nur aus einem ohmschen Widerstand:
\begin{equation}
I(\omega_R)=\frac{U_0}{R}
\end{equation}
Für die Phasenverschiebung von Strom und Eingangsspannung gilt:
\begin{equation}
\tan \varphi=\frac{\omega L - 1 /(\omega C)}{R}
\end{equation}
Der Schwingkreis stellt auch einen Bandpass-Filter dar, die Bandbreite ergibt sich als die Differenz der Frequenzen, bei welchen die Amplitude nur noch das $1/\sqrt{2}$-Fache des Maximus betragen:
\begin{equation}
\Delta \omega=\omega_{1}-\omega_{2}=\frac{R}{L}=2 \delta
\end{equation}

Setzt man nun Kondensator und Spule parallel statt in Reihe, so invertiert sich der BAndfilter und man erhällt eine Bandsperre. Hier ergibt sich als Impendanz:
\begin{equation}
Z_{P}=\left|\frac{1}{\omega L}-\frac{1}{\omega C}\right|
\end{equation}
Damit wird das Verhalten sofort offensichtlich.
