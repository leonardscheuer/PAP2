Ein RC-Glied sei eine Schaltung wie in Abb. 1. ber den Widerstand wird der Kondensator geladen, bis schleilich die Angelegte Gleispannung erreicht wird. Nutzt man $I=\dot{Q}=C\dot{U_C}$ und die Kirchhoffsche Maschenregel, so erhlt man als Lsungen der entstehenden Differentialgleichung:

### Impedanz

Betrachten wir nun einen Wechselstrom, welcher zur Vereinfachung komplexwertig sei (in bekanntem Modell). Dann ist die Impedanz die Verallgemeinerung des Widerstandes als 
\begin{align}
Z=\frac{U}{I}.
\end{align} Fr die hier genutzten Bauteile ergibt sich:
\begin{align}
Z_R&=R\\
Z_C&=-i/(\omega C)\\
Z_L&=i\omega L
\end{align}
Rein imaginre Impedanzen nennt man auch Blindwiderstnde, sie verbrauchen keine Leistung und bewirken nur einen Phasenshift zwischen Strom und Spannung.

### Frequenzverhalten von RC-Gliedern

Legt man an ein RC-Glied nun also eine Wechselspannung an, so addieren sich die Impedanzen wie bei einer normalen Reihenschaltung. ber dem Kondensator erhlt man so die Spannung

\end{align} 
Wodraus sich Betrag und Phase bestimmen lassen:

\end{equation}

### Schwingkreis 
Ein Serienschwingkreis ist eine RCL-Reihenschaltung. Durch die Spule entsteht ein gegenlufiger Strom, wenn der Kondensator sich entldt. Der Widerstand sorgt fr eine Dmpfung, sodass, wie bekannt, das Problem als harmonischer Oszillator beschrieben werden kann. Als allegmeine Lsung findet sich:
\begin{equation}
I(t)=I_{o} e^{-\frac{R}{2 L} t}\left(c_{1} e^{\sqrt{\frac{R^{2}}{4 L^{2}}-\frac{1}{L C} t}}+c_{2} e^{-\sqrt{\frac{R^{2}}{4 L^{2}}-\frac{1}{L C} t}}\right)
\end{equation}
Fr die Resonanzfrequenz ergibt sich: 
\begin{equation}
\omega_R=\sqrt{\frac{1}{LC}}
\end{equation}
Im Resonanzfall verhlt sich der Schwinkreis so, als bestnde er nur aus einem ohmschen Widerstand:
\begin{equation}
I(\omega_R)=\frac{U_0}{R}
\end{equation}
Fr die Phasenverschiebung von Strom und Eingangsspannung gilt:
\begin{equation}
\tan \varphi=\frac{\omega L - 1 /(\omega C)}{R}
\end{equation}
Der Schwingkreis stellt auch einen Bandpass-Filter dar, die Bandbreite ergibt sich als die Differenz der Frequenzen, bei welchen die Amplitude nur noch das $1/\sqrt{2}$-Fache des Maximus betragen:
\begin{equation}
\Delta \omega=\omega_{1}-\omega_{2}=\frac{R}{L}=2 \delta
\end{equation}

Setzt man nun Kondensator und Spule parallel statt in Reihe, so invertiert sich der BAndfilter und man erhllt eine Bandsperre. Hier ergibt sich als Impendanz:
\begin{equation}
Z_{P}=\left|\frac{1}{\omega L}-\frac{1}{\omega C}\right|
\end{equation}
Damit wird das Verhalten sofort offensichtlich.
