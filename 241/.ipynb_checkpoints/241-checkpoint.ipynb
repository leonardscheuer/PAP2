## Grundlagen 
### Zeitentwicklung einer RC-Gliedes

Ein RC-Glied sei eine Schaltung wie in Abb. 1. ber den Widerstand wird der Kondensator geladen, bis schleilich die Angelegte Gleispannung erreicht wird. Nutzt man $I=\dot{Q}=C\dot{U_C}$ und die Kirchhoffsche Maschenregel, so erhlt man als Lsungen der entstehenden Differentialgleichung:
\begin{align}
U_R&=U_Ee^{-t/\tau}\\
I&=I_0e^{-t/\tau}
\end{align}
wobei $\tau=RC$. Experimentell kann $\tau$ bestimmt werden, indem die Halbwertszeit gemessen wird, es ist dann
\begin{align}
\tau=\frac{T_{12}}{\ln 2}
\end{align}

### Impedanz

Betrachten wir nun einen Wechselstrom, welcher zur Vereinfachung komplexwertig sei (in bekanntem Modell). Dann ist die Impedanz die Verallgemeinerung des Widerstandes als 
\begin{align}
Z=\frac{U}{I}.
\end{align} Fr die hier genutzten Bauteile ergibt sich:
\begin{align}
Z_R&=R\\
Z_C&=-i/(\omega C)\\
Z_L&=i\omega L
\end{align}
Rein imaginre Impedanzen nennt man auch Blindwiderstnde, sie verbrauchen keine Leistung und bewirken nur einen Phasenshift zwischen Strom und Spannung.

### Frequenzverhalten von RC-Gliedern

Legt man an ein RC-Glied nun also eine Wechselspannung an, so addieren sich die Impedanzen wie bei einer normalen Reihenschaltung. ber dem Kondensator erhlt man so die Spannung
\begin{align}
U_C=\frac{Z_C}{R+Z_C}U_E
\end{align} 
Wodraus sich Betrag und Phase bestimmen lassen:
\begin{equation}
\begin{array}{c}
\left|U_{C}\right|=\frac{\left|U_{E}\right|}{\sqrt{1+(\omega R C)^{2}}} \\
\tan \varphi=-\omega R C
\end{array}
\end{equation}
Wird stattdessen ber dem Widerstand abgegriffen, so haben wir einen Hochpass gebaut. Analog finden wir Betrag und Phase (siehe Abb. 2). Die enstrechenden Frequenzgnge sind in Abb. 2. sichtbar.

### RC-Glied als Differenzier- und Integrierglied


Abb. 3: Tiefpass als Integrator, Hochpass als Differenziator fr $\tau>>T$. 

RC-Glieder knnen wenn passend gewhlt ($\tau>>T$) (s. auch Abb.3) auch als Differenzier- oder Intergralglied eingesetzt werden. Fr den Hochpass lsst sich dies quantitativ beschreiben als
\begin{equation}
\frac{d U_{A}}{d t}=\frac{1}{R C}\left(U_{E}-U_{A}\right)
\end{equation}
Fr den Tiefpass:
\begin{equation}
U_A=\frac{1}{R C}\int U_{E} dt
\end{equation}

### Schwingkreis 
Ein Serienschwingkreis ist eine RCL-Reihenschaltung. Durch die Spule entsteht ein gegenlufiger Strom, wenn der Kondensator sich entldt. Der Widerstand sorgt fr eine Dmpfung, sodass, wie bekannt, das Problem als harmonischer Oszillator beschrieben werden kann. Als allegmeine Lsung findet sich:
\begin{equation}
I(t)=I_{o} e^{-\frac{R}{2 L} t}\left(c_{1} e^{\sqrt{\frac{R^{2}}{4 L^{2}}-\frac{1}{L C} t}}+c_{2} e^{-\sqrt{\frac{R^{2}}{4 L^{2}}-\frac{1}{L C} t}}\right)
\end{equation}
Fr die Resonanzfrequenz ergibt sich: 
\begin{equation}
\omega_R=\sqrt{\frac{1}{LC}}
\end{equation}
Im Resonanzfall verhlt sich der Schwinkreis so, als bestnde er nur aus einem ohmschen Widerstand:
\begin{equation}
I(\omega_R)=\frac{U_0}{R}
\end{equation}
Fr die Phasenverschiebung von Strom und Eingangsspannung gilt:
\begin{equation}
\tan \varphi=\frac{\omega L - 1 /(\omega C)}{R}
\end{equation}
Der Schwingkreis stellt auch einen Bandpass-Filter dar, die Bandbreite ergibt sich als die Differenz der Frequenzen, bei welchen die Amplitude nur noch das $1/\sqrt{2}$-Fache des Maximus betragen:
\begin{equation}
\Delta \omega=\omega_{1}-\omega_{2}=\frac{R}{L}=2 \delta
\end{equation}

Setzt man nun Kondensator und Spule parallel statt in Reihe, so invertiert sich der BAndfilter und man erhllt eine Bandsperre. Hier ergibt sich als Impendanz:
\begin{equation}
Z_{P}=\left|\frac{1}{\omega L}-\frac{1}{\omega C}\right|
\end{equation}
Damit wird das Verhalten sofort offensichtlich.

### Bestimmung der Zeitkonstante eines RC-Gliedes

Mit den aufgenommenen Halbwertszeiten der RC-Kombinationen werden zunchst die Zeitkonstanten $\tau$ berechent. Es ist
\begin{align}
\tau&=\frac{T_{12}}{\ln 2}\\
\Delta \tau&=\frac{T_{12}}{\ln 2}
\end{align}
Fr den therotischen Wert ist
\begin{equation}
\begin{aligned}
\tau_{t h} &=R C \\
\Delta \tau_{t h} &=\sqrt{(R \cdot \Delta C)^{2}+(C \cdot \Delta R)^{2}}
\end{aligned}
\end{equation}

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from matplotlib.pyplot import figure
import numpy as np
from IPython.display import Markdown, Latex, display
import pandas as pd
from scipy.optimize import curve_fit
import uncertainties
from uncertainties import ufloat as uf
from uncertainties.umath import *
from uncertainties import unumpy as unp
from uncertainties.unumpy import (nominal_values as noms,
                                  std_devs as stds)
from scipy.signal import *

table_counter=1

diag_dpi=100

def round_up(n, decimals = 0):  
    multiplier = 10 ** decimals  
    return np.ceil(n * multiplier) / multiplier 

def out(name,c,unit="",zp=0,decimals=2):
    if isinstance(c, uncertainties.UFloat):
        display(Markdown("$"+name+" = " +'({:L})'.format(c)+" \   \mathrm{"+unit+"}$"))
    elif type(c) in [float,int]:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        print(c)
        
def df_out(df,title=""):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1

def usigma(g1,g2):
    return abs((g1.n-g2.n)/unp.sqrt(g1.s**2+g2.s**2))

def uarr_sigma(g1,g2):
    return abs((noms(g1-g2))/unp.sqrt(stds(g1)**2+stds(g2)**2))
```

```python
v1=pd.read_csv("v1.csv")

v1["C [nF]"]=unp.uarray(v1["C [nF]"],0.1*v1["C [nF]"])
v1["R [k]"]=unp.uarray(v1["R [k]"],0.05*v1["R [k]"])
v1["T [ms]"]=unp.uarray(v1["T [ms]"],v1["T[ms]"])
v1["$\\tau$ [ms]"]=v1["T [ms]"]/np.log(2)
v1["$\\tau_{th}$ [ms]"]=v1["C [nF]"]*v1["R [k]"]*10**(-3)
v1["$\sigma_\\tau$"]=uarr_sigma(v1["$\\tau_{th}$ [ms]"],v1["$\\tau$ [ms]"])
v1=v1.drop(columns=["T[ms]"])
df_out(v1.round(2),"Zeitkonstanten von RC-Gliedern")
```
