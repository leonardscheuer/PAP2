## Grundlagen

### Kräftefreier symmetrischer Kreisel

Ein starrer um einen Festpunkt rotierenden Körper nennen wir einen **Kreisel**. Ist dieser in seinem Schwerpunkt aufgehängt, so ist er **kräftefrei**, da dann kein(gravitatives) Drehmoment entsteht. Der Drehimpuls ist hier also erhalten. Sind zwei der Hauptträgheitsmomente gleich, so nennen wir den Kreisel **symmetrisch**. Einen Kreisel charakterisieren wir anhand drei Achsen:
- **Figurenachse F** 
- **Drehimpulsachse** $\vec{L}$
- **Drehachse** $\vec{\omega}$

Ändert man die an der Figurenachse angreifende Kraft, sodass der Kreiselm (kurz) nicht mehr perfekt Kräftefrei ist, so "taumelt" der Kreisel, er führt eine **Nutation**sbewegung aus. Die Figurenachse bewegt sich dabei um die raumfeste Drehimpulsachse mit $\vec{\omega_N}$ herum, besitzt aber weiterhin eine Eigenrotation $\vec{\omega}_F$. Addieren wir diese Rotationen, so ergibt sich insgesamt:

\begin{equation}
\vec{\omega}=\vec{\omega}_{N}+\vec{\omega}_{F}
\end{equation}

und 

\begin{equation}
\omega_{N}=\frac{L}{I_{x}}
\end{equation}

wobei $I_x$ das entsprechende Trägheitsmoment ist. Für ${\omega_N}$ klein, kann man dies näheren:

\begin{equation}
L \approx I_{z} \omega_{F} \Longrightarrow \omega_{N} \approx \frac{I_{z}}{I_{x}} \omega_{F}
\end{equation}

Für die Winkelgeschwindigkeit $\Omega$ des Körperkegels ergibt sich:
\begin{equation}
\Omega=\frac{I_{x}-I_{z}}{I_{x}} \omega_{F}
\end{equation}
Es lässt sich
\begin{equation}
I_{x}-I_{z}=\frac{I_{z}}{\omega_{F} / \Omega-1}
\end{equation}
zeigen.

### Der schwere, symmetrische Kreisel
Der Kreisel sei nun nicht mehr im Schwerpunkt, aber immernoch auf der Figurenachse aufgehängt. Es ergibt sich also ein Drehmoment durch die am Schwerpunkt angreifende Kraft:
\begin{equation}
\vec{M}=\vec{l} \times m \vec{g}
\end{equation}

Wobei $\vec{l}$ vom Unterstützungspunkt zum Schwerpunkt reicht. Der Drehimpuls ist nun also Änderung unterworfen. Der Drehimpulsvektor bewegt sich nun in einem Kegel um die z-Achse herum, dies nennen wir **Präzession**. $\vec{\omega_P}$ sei die Präzessionsfrequenz, mit welcher $\vec{L}$ um die z-Achse rotiert. Man findet:
\begin{equation}
\omega_{P}=\frac{m g l}{I_{z} \omega_{F}}
\end{equation}
sodass allgemein gilt:
\begin{equation}
\vec{M}=\overrightarrow{\omega_{P}} \times \vec{L}
\end{equation}

### Dämpfung des Kreisels
Für die Dämpfungskonstante $\delta$ gilt:
\begin{equation}
\omega(t)=\omega_{0} e^{-\delta t}
\end{equation}
Wir tragen $\omega$ daher logarithmisch auf um $\delta$ zu erhalten: 

```python
#Funktionen zur späteren Verwendung. 
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
from IPython.display import Markdown, display, Latex
import pandas as pd
from scipy.optimize import curve_fit
from uncertainties import ufloat
from uncertainties.umath import *
from uncertainties import unumpy as unp
from uncertainties.unumpy import (nominal_values as noms,
                                  std_devs as stds)

table_counter=1

def comma_to_float(valstr):
  return float(valstr.decode("utf-8").replace(',','.'))

def round_up(n, decimals = 0):  
    multiplier = 10 ** decimals  
    return np.ceil(n * multiplier) / multiplier 

def outwstd(name,c,Dc,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = (" +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+") "+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(Dc*10**(-zp),decimals),decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))

def out_sys_std(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        pass
        display(Markdown( "$ "+name+" = ( " +"{:.{}f}".format(round(c[0]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+ "{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals) +" ) " + unit + " $"))
    else:
        display(Markdown( "$ "+name+" = ( " + "{:.{}f}".format(round(c[0]*10**(-zp),decimals),decimals)+" \pm "+"{:.{}f}".format(round_up(c[1]*10**(-zp),decimals), decimals)+" \pm "+"{:.{}f}".format(round_up(c[2]*10**(-zp),decimals), decimals)+" )10^{ "+str(zp)+" }  " + unit + " $"))
        
def out_no_error(name,c,zp=0,decimals=0,unit=""):
    if zp==0:
        display(Markdown("$"+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+unit+"$"))
    else:
        display(Markdown( "$ "+name+" = " +"{:.{}f}".format(round(c*10**(-zp),decimals),decimals)+" \cdot 10^{ "+str(zp)+" }  " + unit + " $"))

def df_out(df,title):
    global table_counter
    display(Markdown("**Tabelle "+str(table_counter)+": " +title+ "**"))
    display(Markdown(df.to_markdown()))
    table_counter+=1
    
def sigma(g1,Dg1,g2,Dg2):
    return abs((g1-g2)/np.sqrt(Dg1**2+Dg2**2))

def quad_add(arr):
    quad_sum=0
    for i in arr:
        quad_sum+=i**2
    return (quad_sum**0.5)/len(arr)

def fit_line(x,a,b):
    return a*x+b

def fit_propline(x,a):
    return a*x
```

```python
t, omp, omp_err = np.loadtxt('daempfung.txt', unpack = True, skiprows = 1)

om=unp.uarray(omp,omp_err)

om *= 2 * np.pi / 60
t_err = np.zeros(len(t)) + 1

popt, pcov = curve_fit(fit_line, t, unp.log(noms(om)), sigma = stds(om)/noms(om)) #Gauß, t_err neglegible

x = np.linspace(0, 720, 2)
plt.plot(x, fit_line(x, popt[0], popt[1]), label = 'Ausgleichsgerade')
plt.legend()

plt.errorbar(t, np.log(noms(om)), xerr = t_err, yerr = stds(om)/noms(om), fmt = 'bx', capsize = 3)

plt.xlabel('t [s]')
plt.ylabel('$\ln(\omega)$ [\"ln(Hz)\"]')
plt.title('Winkelgeschwindigkeit')
plt.grid(ls='dotted')
delta = -popt[0]
delta_err = pcov[0][0] #Gauß
outwstd("\\delta",delta, delta_err,-4,7,"\\frac{1}{s}")
```

Wir bestimmen nun die Halbwertszeit mit: 
$\newcommand{\lr}[1]{\left(#1\right)}$
\begin{equation}
t_{1 / 2}=\frac{\ln (2)}{\delta}
\end{equation}
\begin{align}
	\Delta t_{1/2} = \ln(2)\frac{\Delta \delta}{\delta^2}
\end{align}

```python
t_hw = np.log(2)/delta
t_hw_err = np.log(2)*delta_err/(delta**2)
outwstd("t_{1/2}",t_hw, t_hw_err,0,5,"s")
```

### Interpretation der Präzessionsdauern bei verscheidenen Startwinkeln
Wir wollen nun die gemessenen Präzessionsdauern diskutieren. Theoretisch erwarten wir dafür:
\begin{align}
	\omega_P&=\frac{I_F\omega_F}{mgl}\\
	\therefore  T&=\frac{2\pi I_F}{mgl}\omega_F =\frac{4\pi^2 I_F}{mgl}f
\end{align}

Insbesondere erwarten wir also $f \propto T$.
Betrachten wir diesen Zusammenhang, so ist dies schwer erkennbar:

```python
f1, winkel, T1 = np.loadtxt('praezession_a.txt', unpack = True, skiprows = 1)
winkel_err=np.ones(len(winkel))*5
f1 = f1/60 #SI units
f1_err = (np.zeros(len(f1))+2)/60
T1_err = np.zeros(len(T1)) + 5

plt.errorbar(f1, T1, xerr = f1_err, yerr = T1_err, fmt = 'bx', capsize = 3)

popt, pcov = curve_fit(fit_propline, f1, T1, sigma = T1_err)
plt.plot(x, fit_propline(x, popt[0]), label = 'Ausgleichsgerade')
plt.legend()
plt.grid(ls='dotted')

plt.xlabel('$f$ [Hz]')
plt.ylabel('$T$ [s]')
_=plt.title('Präzessionsdauer in Abhängigkeit von der Kreisfrequenz')


```

### Bestimmung des Trägheitsmoments 
$\renewcommand{\t}[1]{\text{#1}}$
Wir wollen nun mit (13) $I_F$ bestimmen. Ist $k$ die Steigung in dem nachfolgenden Diagramm, so ergibt sich:
\begin{align}
	& k = \frac{2\pi I_F}{gml}\\
	\Leftrightarrow & I_F = \frac{k\cdot gml}{2\pi}\\
	\stackrel{\t{Gauß}}\Rightarrow & \Delta I_F = I_F\sqrt{\lr{\frac{\Delta k}{k}}^2+\lr{\frac{\Delta l}{l}}^2}
\end{align}
wobei wir $\omega_F$ durch die Anfangsfrequenz und den Abfall durch die Dämpfung beschreiben:
\begin{align}
	&\bar \omega_F = \frac{1+e^{-\delta T_p}}{2} \omega_{F,0} = (1+e^{-\delta T_p})\pi\cdot f_0\\
	 & \Delta \bar \omega_F = \bar \omega_F \sqrt{\lr{\frac{\Delta f_0}{f_0}}^2+\lr{\frac{e^{-\delta T_P}\cdot T_P}{1+e^{-\delta T_P}}\Delta \delta}^2}
\end{align}
Wir mitteln schließlich noch über alle vier Messreihen und erhalten:

```python
f, T = np.loadtxt('praezession_b.txt', unpack = True, skiprows = 1, comments = '>')
f = f/60 
f_err = (np.zeros(len(f))+2)/60
om_bar = (1+np.exp(- delta * T)) * np.pi * f
om_bar_err = om_bar * np.sqrt((f_err/f)**2 + (np.exp(- delta * T) * T * delta_err / (1+np.exp(- delta * T)))**2)

T_err = np.zeros(len(T)) + 5

k = np.zeros(4)
k_err = np.zeros(4)
x = np.linspace(20, 70, 2)

Messreihe=["Ein Gewichstsstück bei 15cm","Ein Gewichstsstück bei 20cm","Zwei Gewichstsstücke bei 15cm","Zwei Gewichstsstücke bei 20cm"]

for i in range(0,4):
    plt.errorbar(om_bar[4*i:4*(i+1)], T[4*i:4*(i+1)], xerr = om_bar_err[4*i:4*(i+1)], yerr = T_err[4*i:4*(i+1)], fmt = 'x', capsize = 3, label = Messreihe[i])
    popt, pcov = curve_fit(fit_propline, om_bar[4*i:4*(i+1)], T[4*i:4*(i+1)], sigma = T_err[4*i:4*(i+1)])
    k[i] = popt[0]
    k_err[i] = pcov[0][0]**0.5
    plt.plot(x, fit_propline(x, k[i]), 'b')

plt.xlabel('$\omega_F$ [Hz]')
plt.ylabel('$T$ [s]')
plt.title('Präzessionsdauer in Abhängigkeit von der Kreisfrequenz')
plt.grid(ls='dotted')
_=plt.legend()

l = np.array([0.15, 0.2, 0.15, 0.2])
l_err = np.zeros(4)+0.2*10**(-3)
m = np.array([1,1,2,2]) * 9.85 * 10**(-3)
I_F = k * 9.80984 * m * l / (2 * np.pi)
I_F_err = I_F * np.sqrt((l_err/l)**2 + (k_err/k)**2)
I_F_final = np.mean(I_F)
I_F_final_err = np.sqrt(np.sum(I_F_err**2))/len(I_F_err)
outwstd("I_F",I_F_final, I_F_final_err,-3,3,"kgm^2")
```

Wir vergleichen dies mit dem theoretischen Trägheitsmoment der Kugel:
\begin{align}
	I = \frac 25 mr^2 
\end{align}

```python
I_theo=(2/5)*4.164*0.0508**2
out_no_error("I_{theo}",I_theo,zp=-3,decimals=2,unit="kgm^2")
out_no_error("\sigma",sigma(I_theo,0,I_F_final,I_F_final_err),zp=0,decimals=2,unit="")
```

Man findet hier also eine gute Übereinstimmung.

### Bestimmung von $I_\perp$ über Umlauffrequenz $\Omega$

Wir wollen nun das Trägheitsmoment orthogonal zur Figuenachse ermitten. Dafür verwenden wir (4) und erhalten bei Auftragung von $\Omega$ gegen $\omega_F$ die Steigung
\begin{align}
	&m = \frac{I_\perp-I_F}{I_\perp}\\
	\therefore & I_\perp = \frac{I_F}{1-m}\\
	\therefore & \Delta I_\perp = I_\perp\sqrt{\lr{\frac{\Delta I_F}{I_F}}^2+\lr{\frac{\Delta m}{1-m}}^2}
\end{align}

```python
om2, om_T = np.loadtxt('momDrehachse.txt', unpack = True, skiprows = 1)

om2 *= 2 * np.pi / 60 #winkelgeschwindigkeit
om2_err = np.zeros(len(om2)) + 10 * 2 * np.pi / 60
om_T_err = 20 * np.pi / om_T**2 * 0.5
om_T = 20 * np.pi / om_T #winkelgeschwindigkeit

plt.errorbar(om2, om_T, xerr = om2_err, yerr = om_T_err, fmt = 'bx', capsize = 3)

popt, pcov = curve_fit(fit_propline, om_T, om2, sigma = om2_err) #om_T_err neglegible

x = np.linspace(30, 63, 2)

plt.plot(x, fit_propline(x, 1/popt[0]), 'b', label = 'Ausgleichsgerade')

plt.xlabel('$\omega_F$ [Hz]')
plt.ylabel('$\Omega$ [Hz]')
plt.title('Umlauffrequenz in Abhängigkeit von der Kreisfrequenz')
plt.grid(ls='dotted')
plt.legend()
I_perp = I_F_final/(1-1/popt[0])
I_perp_err = I_perp * np.sqrt((I_F_final_err/I_F_final)**2 + (pcov[0][0]/(popt[0]**2*(1-1/popt[0])))**2)
outwstd("I_\\perp",I_perp, I_perp_err,-3,3,"kgm^2")
```

### Bestimmung von $I_\perp$ über Nutation
Jetzt nutzen wir (3) um $I_\perp$ zu bestimmen:
\begin{align}
	& m \frac{\omega_N}{\omega_F} = \frac{I_F}{I_\perp}\\
	\therefore & I_\perp = \frac{I_F}{m}\\
	\therefore & \Delta I_\perp = I_\perp \sqrt{\lr{\frac{\Delta I_F}{I_F}}^2+\lr{\frac{\Delta m}{m}}^2}
\end{align}

```python
om_F, om_N = np.loadtxt('nutation.txt', unpack = True, skiprows = 1)
om_F *= 2 * np.pi / 60
om_N *= 2 * np.pi / 60 #winkelgeschwindigkeit
om_F_err = np.zeros(len(om2)) + 10 * 2 * np.pi / 60
om_N_err = np.zeros(len(om2)) + 10 * 2 * np.pi / 60

plt.errorbar(om_F, om_N, xerr = om_F_err, yerr = om_N_err, fmt = 'bx', capsize = 3)

popt, pcov = curve_fit(fit_propline, om_F, om_N, sigma = np.sqrt(om_N_err**2+(I_F_final/I_perp*om_F_err)**2))

x = np.linspace(55, 85, 2)

plt.plot(x, fit_propline(x, popt[0]), 'b', label = 'Ausgleichsgerade')

plt.xlabel('$\omega_F$ [Hz]')
plt.ylabel('$\omega_N$ [Hz]')
plt.title('Nutationsfrequenz in Abhängigkeit von der Kreisfrequenz')
plt.grid(ls='dotted')
plt.legend()

I_perp2 = I_F_final/popt[0]
I_perp2_err = I_perp2 * np.sqrt((I_F_final_err/I_F_final)**2 + (pcov[0][0]/popt[0])**2)
outwstd("I_\\perp",I_perp2, I_perp2_err,-3,3,"kgm^2")
```

Wir vergleichen die Ergebnisse:

```python
out_no_error("\sigma",sigma(I_perp,I_perp_err,I_perp2,I_perp2_err),zp=0,decimals=2,unit="")
```

## Diskussion
In diesem Versuch haben wir uns zunächst im Vorversuch ein Gefühl für das Verhalten eines Kreisels verschafft. Anschließend haben wir duch Messung über 12 Minuten die Dämpfung des Kreisels bestimmt. Der Fehler der Dämpfungskonstanten ist sehr klein (6 Größenordnungen kleiner als die Größe selbst). Danach wollten wir die Präzessionseigenschaften des Kreisels untersuchen. Wir hätten theoretisch erwartet, dass die Präzessionsdauern unabhängig vom Winkel sind, dies konnte jedoch nicht verifiziert werden. Die Fehler sind jedoch so groß, dass es auch nicht als widerlegt gesehen kann. Vermutlich liegen hier bei zumindest einem der Messwerte ein Messfehler vorliegt. Wir haben schließlich das Trägheitsmoment um die Figurenachse bestimmt, der ermittelte Wert stimmt äußerst gut mit dem theoretischen Wert überein ($<0.05\sigma$). Schließlich haben wir das Trägheitsmoment orthogonal zur Figurenachse auf zwei verschiedene Arten bestimmt. Zuerst haben wir dafür die Umlauffrequenz $\Omega$ genutzt, dann die Nutation. Beide Werte unterscheiden sich Signifikant ($>20\sigma$). (Mindestens) eine der Messungen scheint einen systematischen Fehler aufzuweisen. Errechnet man das theorische Trägheitsmoment, so erhält man einen nahezu identischen Wert zur ersten Messung ($4,56 \cdot 10^{-3} kgm^2$, Abweichung $<0.5\sigma$). Das deutet darauf hin, dass der Fehler bei der Messung der Nutationsfrequenzen auftrat. Die Aufnahme der Messwerte mit dem Stoboskop könnte einen systematischen Fehler aufweisen. Auch wurde eine kleinwinkelnäherung in der zugrundeliegenden Theorie verwendet, eventuell war der Nutationskegel während der Messung zu weit geöffnet, damit diese ihre vollständige Gültigkeit behält.
